## Theorem 2:

$\partial \Omega \in C^2 \rightarrow \partial \Omega$ satisfies the exterior ball condition.
$\forall \zeta \in \partial \Omega, \exists B_R(x_0') \subset \bar{\Omega}^C, B_R(x_0') \cap \bar{\Omega} = \{ \zeta \}$

Then

$$
w_\zeta =
\begin{cases}
-\frac{1}{|x - x_0'|^{n - 2}} + \frac{1}{R^{n - 2}}, & n \geq 3\\
\ln \left(\frac{|x - x_0'|}{R} \right), & n = 2
\end{cases}
$$

$\partial \Omega$ satisfies the exterior cone condition; if $\forall \zeta \in \partial \Omega, \exists$ a cone $K_\zeta$ centered at $\zeta$ s.t. $K_\zeta \cap \bar{\Omega}^C, \bar{K}_\zeta \cap \bar{\Omega} = \{ \zeta \}$

Let's try to construct a barrier for $n = 2, \zeta = 0$: $w = r^\lambda \cos(\mu \theta)$ where $0 < \theta < \theta_0 < 2 \pi$.

$\theta_0$ is the angle of the cone

Then

$$
\Delta w = w_{rr} + \frac{1}{r} w_r + \frac{1}{r^2} w_{\theta \theta} = r^{\lambda - 2} \cos(\mu \theta) (\lambda^2 - \mu^2)
$$

Then we can take $\lambda = \pm \mu$ to make this a harmonic function.

Make sure $w > 0$ in $\bar{\Omega} \setminus \{ \zeta \}$, with $0 < \lambda = \mu << 1$
Then $w > 0$ is always positive in a local region $\bar{B}_R(\zeta) \cap \bar{\Omega} \setminus \{ \zeta \}$.

The barrier is a slit in the domain

For $n \geq 3$, take a polar coordinate at $\zeta = 0$ (wlog), then $x = r \vec{\theta}, \theta \in S^{n - 1} \setminus \Sigma$ where $\Sigma$ is some set opposite from the cone.

On $S^{n - 1} \setminus \Sigma = M$, we can do a Dirichlet boundary condition, with

Then use

We only need a small cone, then this can be the local variable

## Summary
(1)
$$
-\Delta u = f, \in \Omega\\
u = g, \in \partial \Omega
$$

We proved that if $f \in C^1(\bar{\Omega}), g \in C^0(\partial \Omega)$, and $\partial \Omega$ is Lipschitz, then $\exists$ a solution, with

$$
u(x) = \int_{\Omega} G(x, y) f(y) dy - \int_{\partial \Omega} g \frac{\partial G}{\partial \nu} d \sigma_y
$$

Can we find a solution with lower regularity requirements? Yes, the Dirichlet Principle:

Associated with (1), we have the following energy function
$$
E[u] = \frac{1}{2} \int_\Omega |\nabla u|^2 - \int_\Omega f g
$$

Admissible Set
$$
A = \{ u, \in C^2(\bar{\Omega}) | u = g, \in \partial \Omega \}
$$

Consider the following minimization problem:
$c = \inf_{u \in A} E[u]$

We don't know that the infinum exists in the admissible set, so we can't say $\min$

## Theorem 2.2.20 (Dirichlet principle)
(a) If a solution $u_0 \in A$ satisfying (1) exists, then $E[u_0] = c \leq E[u] \forall u \in A$

(b) If $c$ is attained by some $u_0 \in A$, then $u$ satisfies (1)

Note this implies some stronger differentiability of $g$

The Dirichlet principle provides a way of obtaining a solution by minimization of the energy $E[u]$.

### Proof
(a)
$$
\begin{cases}
-\Delta u_0 = f, & \in \Omega\\
u_0 = g, & \in \partial \Omega\\
\end{cases}
u \in A \rightarrow u - u_0 = 0 \in \partial \Omega\\
-\Delta u_0 = f \times (u - u_0)\\
-\int (u - u_0) \Delta u_0 = \int f (u - u_0) = \int \nabla u_0 \cdot \nabla (u - u_0)\\
\rightarrow \int \nabla u_0 \cdot \nabla (u - u_0) - \int f (u - u_0) = 0\\
E[u] = E[u + u - u_0] = \frac{1}{2} \int |\nabla u_0 + \nabla (u - u_0)|^2 - \int (u_0 + u - u_0) f\\
= E[u_0] + \int \nabla u_0 \cdot \nabla (u - u_0) - \int f(u - u_0) + \frac{1}{2} \int |\nabla (u - u_0)|^2\\
= E[u_0] + \frac{1}{2} \int | \nabla (u - u_0)|^2
$$
proving (a)

(b)
$$
\forall \phi \in C_0^\infty(\Omega)\\
E[u_0 + \epsilon \phi] = E[u_0] + \epsilon \left( \int \nabla u_0 \cdot \nabla \phi - \int f \phi \right) + \frac{\epsilon^2}{2} \int |\nabla \phi|^2
$$
attains a minimum at $\epsilon = 0$
Then
$$
\int \nabla u_0 \cdot \nabla \phi - \int f \phi = 0 \rightarrow \int_\Omega (-\Delta u_0 - f) \phi = 0
$$
