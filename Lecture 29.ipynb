# Regularity of Weak Solutions
$$
\begin{cases}
L u = f & \in \Omega\\
u = g & \in \partial \Omega
\end{cases}\\
$$

$u \in W_0^{1, 2}(\Omega)$ is a weak solution $\leftarrow \rightarrow \int_\Omega a^{i, j} \partial_i \partial_j + b_j \partial_j u v + c u v dx$

## Theorem 1
$$
||u||_{W^{2, 2}(V)} \leq C (||u||_{L^2(\Omega)} + ||f||_{L^2(\Omega)})
$$

### Remark
1) From local $H^2$ estimates, we know $L u \in L_{loc}^2$
In the definition of the weak solution, $v \in C_0^\infty(\Omega)$

$\int_\Omega (L u) v dx = \int_\Omega a^{i, j} \partial_i u \partial_j v + \ldots = \int_\Omega f v$
$\rightarrow L u = f$ almost everywhere

2)
$$
\begin{cases}
\Delta u + \lambda u = 0 & \in \Omega & \lambda_1 < \lambda_2 < \lambda_3 < \ldots < \lambda_n < \ldots\\
u = 0 & \in \partial \Omega & \lambda_n \rightarrow \infty
\end{cases}\\
u \in H_0^1(\Omega) \rightarrow \text{iteration} \rightarrow -\Delta u = \lambda u \in L^2\\
\rightarrow u \in H_{loc}^2(\Omega) \rightarrow u \in H_{loc}^m \forall m \geq 2 \rightarrow u \in C^{\infty}(\Omega)
$$

Global $H^2$ estimates

## Theorem 2
Given $a^{i, j} \in C^1, b_j, c \in L^\infty, \partial \Omega \in C^2$, if $u$ is a weak solution to
$$
\begin{cases}
L u = f & \in \Omega\\
u = 0 & \in \partial \Omega
\end{cases}\\
||u||_{W^{2, 2}(\Omega)} \leq C (||u||_{L^2(\Omega)} + ||f||_{L^2(\Omega)})
$$

### Proof
1) Model Case
2) Change of variable, partition of unit

Model Case: $\Omega = B_1^+(0) = B_1(0) \cap \{x_1 > 0\}$

$\eta = 1$ in $\bar{B}_{\frac{1}{2}}^+$
$\eta = 0$ in $\bar{B}_1^+ \setminus \bar{B}_{\frac{3}{4}}^+$
$v = \partial_k^{-h}(\eta^2 \partial_k^h u), k = 1, 2, \ldots, n - 1$

#### Claim: $v \in H_0^1(\Omega)$
$u(x', 0) = 0$
$$
v(x) = \partial_k^{-h}\left(\eta^2(x) \frac{u(x + h e_k) - u(x)}{h}\right)
  = \frac{1}{-h} \left[ \eta^2(x - h e_k) \frac{u(x - h e_k + h e_k) - u(x - h e_k)}{h} - \eta^2(x) \frac{u(x + h e_k) - u(x)}{h}\right]\\
x \in \partial B_1^+ \setminus \{ x_n = 0 \} \rightarrow \eta(x) = 0, \eta(x - h e_k) = 0, v = 0\\
x \in \bar{B}_1^+ \cap \{ x_n = 0 \} \rightarrow x_n = 0, u(x - h e_k) = u(x' - h e_k, 0) = 0, v = 0
$$

Then repeat the proof from the previous lecture:
$$
\int a^{i, j} \partial_i u \partial_j (\partial_k^{-h}(\eta^2 \partial_k^h v))
  = \int (f - b_j \partial_j u - c u) \partial_k^{-h} (\eta^2 \partial_k^h v))\\
\int \eta^2 |\partial_k^h D v|^2 \leq c \int_\Omega |f|^2 + \int_\Omega |D u|^2 + \int_\Omega u^2\\
  \rightarrow \int_{B_{\frac{1}{2}}^+} \eta^2 |\partial_k D u|^2 \leq C
  \rightarrow \sum_{i + j \leq 2 n - 1} \int_{B_{\frac{1}{2}}} |\partial_i \partial_j u|^2 \leq C
$$
So after the same proof as previously, we're only missing bounds on $\partial_n \partial_n u$
Recall $L u = f$ almost everywhere
Then
$$
-\partial_i (a^{i, j} \partial_j u) = -b_k \partial_j u - c u + f \in L^2\\
-a^{n n} \partial_n (\partial_n u) = \sum_{i + j \leq 2 n - 1} \partial_i(a^{i, j} \partial_j u) \in L^2\\
-\partial_n \partial_ u = \frac{1}{a^{n, n}} [ \ldots ] \in L^2
$$

This proves the model case; next we show the same is true for the general case

In the general case, take $x_0 \in \partial \Omega$
Then with the maps $\Phi \in C^2$ from $\Omega$ to $B_1^+$ and $\Psi \in C^2$ as its inverse, consider $\tile{u}(y) = u(\Psi(y)): B_1^+$

$$
\int a^{i, j}(x) \partial_j u(x) \partial_i v(x) + b_j \partial_j u(x) v + c u v = \int f v\\
x = \Psi(y), y = \Phi(x), \frac{\partial}{\partial x_j} = \frac{\partial \Phi_k}{\partial x_j} \frac{\partial}{\partial y_k}\\
\int |D I| a^{i, j}(\Psi(y)) \frac{\partial \Phi_k}{\partial x_j} \frac{\partial \Phi_l}{\partial x_i} \frac{\partial u}{\partial y_k} \frac{\partial v}{\partial y_l} dy\\
\tilde{a}^{k, l}(y) = a^{i, j} \frac{\partial \Phi_k}{\partial x_j} \frac{\partial \Phi_l}{\partial x_i}\\
1) \tilde{a}^{k, l} \in C^1\\
2) \tilde{a}^{k, l} \zeta_k \zeta_l = a^{i, j} \frac{\partial \Phi_k}{\partial x_j} \frac{\partial \Phi_l}{\partial x_i} \zeta_k \zeta_l\\
  = a^{i, j} \left( \frac{\partial \Phi_k}{\partial x_j} \zeta_k \right) \left( \frac{\partial \Phi_l}{\partial x_i} \zeta_l \right)\\
  \geq \theta \sum_j \sum_k \left| \frac{\partial \Phi_k}{\partial x_j} \zeta_k \right|^2
  = \theta \sum_k \left( \sum_j \left| \frac{\partial \Phi_k}{\partial x_j} \right|^2 \right) |\zeta_k|^2\\
  \geq C^1 \theta |\zeta|^2
$$

## $H^2$ estimates

(Local): $||u||_{W^{2, 2}(V)} \leq c (||u||_{L^2(\Omega)} + ||f||_{L^2(\Omega)})$

(Global): $||u||_{W^{2, 2}(\Omega)} \leq c (||u||_{L^2(\Omega)} + ||f||_{L^2(\Omega)})$
whenever $a^{i, j} \in C^1, b_j \in L^\infty, c \in L^\infty$

## Summary of Regularity Theorems
1) $H^2$ estimates, local and global
2) $L u = f \in L^p(\Omega), p > 1$
$W^{2, p}$ Theory (Calderon-Zygmond Theory)

Local $L^p$ theory: $1 < p < \infty$
$$
||u||_{W^{2, p}(V)} \leq c (||u||_{L^p(\Omega)} + ||f||_{L^p(\Omega)}) \forall V \subset \subset \Omega
$$

Global $L^p$ theory: $1 < p < \infty$
$$
||u||_W^{2, p}(\Omega) \leq C (||u||_{L^p(\Omega)} + ||f||_{L^p(\Omega)}), a^{i, j}, b^j, c \in L^\infty, \partial \Omega \in C^1
$$
Beyond the scope of this course, see Gilbarg and Trudinger's book
$p > \frac{n}{2}, W^{2, p}(\Omega) \rightarrow C^{2 - \frac{n}{p}}, u \in C^\alpha$

3) Schauder-Theory
$L u = f \in C^\alpha$
