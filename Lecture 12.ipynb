(Today)
If we assume that $|u(x, t)| \leq c e^{a |x|^2}, \forall t > 0$, then $u$ is given by (2).

### Parabolic Maximum Principle
$u_t - \Delta u \leq 0$ in $\Omega \times (0, T), \Omega$ bounded

Then $\max_{\bar{\Omega \times (0, T)}} u(x, t) = \max_{\partial'(\Omega \times (0, T))} u(x, t)$

### Proof of Theorem 2
Let $v(x, t)$ be a solution to (1). Let $w = u - v$.

$$
w_t = \Delta w\\
|w| \leq e^{a |x|^2}\\
w(x, 0) = 0
$$

#### Claim:
$w(x, t) \leq 0, \forall x, t$

#### Proof
$\Omega = B_R(0), T > 0$
Choose $\epsilon > 0$ s.t. $4 (T + \epsilon) < 1$

Let $\tilde{w}(x, t) = w(x, t) - \delta (4 \pi (T + \epsilon - t))^{-n/2} e ^{\frac{|x|^2}{4(T - \epsilon - t)}}$

Can check that this satisfies the heat equation, substiture $\epsilon = 0$, then $v_t = \Delta v$

Then $\tilde{w}_t = \Delta \tilde{w}$

By the parabolic maximum principle, we have $\max_{\bar{\Omega \times (0, T)}} \tilde{w}(x, t) = \max_{\partial'(\Omega \times (0, T))} \tilde{w}(x, t)$.

On $B_R \times \{ 0 \}$, we have $\tilde{w} \leq 0$
On $B_R \times (0, T]$, we have $|x| = R$
Then

$$
\tilde{w}(x, t) = w(x, t) - \delta (4 \pi (T + \epsilon - t))^{-n / 2} e^{\frac{R^2}{4(T + \epsilon - t)}}\\
\leq c e^{a R^2 - \delta (4 \pi (T + \epsilon))^{-n / 2} e^{\frac{R^2}{4(T + \epsilon)}}
$$

Choose $R \geq R_0(\delta, \epsilon$, so that $\tilde{w}(x, t) \leq 0$.

Then $a < \frac{1}{4(T + \epsilon)}$
and $\tilde{w}(x, t) \leq 0, |x| \geq R_0$,
so $w(x, t) \leq \delta ((4 \pi(T + \epsilon - t))^{-n / 2} e^{\frac{|x|^2}{4 (T + \epsilon - t)}}$

Take $R$ large enough and $\tilde{w}(x, t) \leq 0$ is true for all $x$ with $0 < t < T$

Finally, let $\delta \rightarrow 0+$.
Then $w(x, t) \leq 0 \forall x, 0 < t < T$.

Change $w$ to $-w$, so $-w(x, t) \leq 0, \forall x, 0 < t < T$, so $w(x, t) = 0$
$w(x, t) = 0, \forall t > 0$

This proves uniqueness.

We've proven uniqueness and the representation formula.

## Theorem 3 (Notes Tychonov's Example)

$\exists$ an non-trivial solution to the homogeneous heat equation.

$$
\begin{cases}
w_t = \Delta w, & t > 0\\
w(x, 0) = 0
\end{cases}
$$

such that $|w| \leq e^{\frac{a |x|^2}{t}}$

Non-uniqueness of solutions to the heat equations.

### Proof
Fix $t$, consider it as some kind of parameter. Expand $u(x, t) = \sum_{j=0}^{\infty} g_j(t) x^j$.
Assume that $u$ is even by construction, so $u_x(0, t) = 0$

$u_t = \sum_{j = 0}^\infty g_j'(t) x^j$, $u_{xx} = \sum_{j = 0}^\infty j (j - 1) g_{j}(t) x^{j - 2}$
Then $g_j'(t) = (j + 1)(j + 2) g_j(t)$

$$
g_0(t) =
\begin{cases}
e^{-1/t^\alpha}, & t \geq 0\\
0, & t = 0
\end{cases}
\alpha > 1\\
g_1 = 0\\
g_{j + 2}(t) = \frac{1}{(j + 1)(j + 2)} g_j'(t)\\
g_{2k}(t) = \frac{1}{(2 k)!} g^{(k)}(t)\\
u(x, t) = \sum_{k = 0}^\infty \frac{1}{(2 k)!} g^{(k)}(t) x^{2 k}\\
|g^{(k)}(t)| \leq \frac{k!}{(\theta t)^k} e^{-\frac{1}{t^\alpha}}
$$

Then
$$
|u(x, t)| \leq \sum_{k = 0}^\infty \frac{k!}{(2 k)!} \frac{1}{(\theta t)^k} |x|^{2 k} e^{-\frac{1}{t^\alpha}}\\
\leq \sum_{k = 0}^\infty \frac{1}{k!} \frac{1}{(\theta t)^k} |x|^{2 k} e^{-\frac{1}{t^\alpha}}
$$

$$
= e^{\frac{|x|^2}{\theta t} - \frac{1}{t^\alpha}}
$$

So $u, u_t, u_xx$ converges uniformly when $t$ is fixed.

But as $t \rightarrow 0+$, we have $|u| \leq e^{-\frac{1}{1000000 \theta t^\alpha}}$
(or any constant $>1$ rather than just $1000000$)
