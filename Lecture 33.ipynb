## Theorem 1 (Strong Maximum Principle)
$$
\begin{align*}
L u \geq 0, c(x) \leq 0 & \in \Omega
\end{align*}
$$
Then $u$ can't attain it's nonnegative maximum inside the domain unless it's constant

### Proof
$M = \max_{\bar{\Omega}} u$ Assume that $M \geq 0$.
Suppose $M$ is attained by some interior point
Then
$\Sigma = \{ x \in \bar{\Omega} | u(x) = M \}$ is closed (why?), and $\Sigma \cap \Omega^0 \neq \emptyset$
If $\Sigma = \bar{\Omega}$, we are done. Othewise, consider $\Omega \setminus \Sigma$
Remove a small ball from the remainder until you touch $\Sigma$
$$
0 \leq u(x_0) = M, u(x) < u(x_0) \hspace{5mm} \in \Omega \setminus \Sigma
$$
By the Hopf boundary lemma, with $n$ the normal on the boundary of $\Omega \setminus \Sigma$
$$
\frac{\partial u}{\partial n}(x_0) > 0
$$
Then we have a contradiction, as $M$ is not the maximum

## Defintion
Let $L u = a^{i, j} u_{x_i x_j} + b_j u_j + c(x) u$
We say that $w$ is a super-solution to $L$ if

1) $L w \leq 0$ in $\Omega$

2) $w > 0$ in $\Omega$

### Example
$c(x) \leq 0$, then $w = 1$ is a super-solution to $L$

## Generalized Hopf Boundary Lemma
Suppose $L u \geq 0$ in $B$ and $\exists$ a super-solution $w(x)$.
Suppose
$$
\frac{u(x)}{w(x)} < \frac{u(x_0)}{w(x_0)}
$$
Then $\left. \frac{\partial}{\partial n} \left( \frac{u}{w} \right) \right\rvert_{x=x_0} > 0$

### Proof
$$
u(x) = w v(x)\\
a^{i, j} v_{x_i x_j} + (\tilde{b}_j) v_j + \left( \frac{L w}{w} \right) v \geq 0
$$

## Theorem 4: Narrow Domain Principle
If your domain is in a small stream, you can always use the maximum principle

Let $L u = a^{i, j} u_{x_i x_j} + b_j u_j + c(x) u$
$$
\Omega \subset \{ 0 < x_1 < d \}
$$
Then $\exists d_0 > 0$ s.t. for any $d < d_0$, $L$ has a super-solution; i.e., $L$ satisfies the strong maximum principle
$$
w(x) = e^{\alpha d} - e^{\alpha x_1} > 0\\
L w = -(a^{1, 1} \alpha^2 + b_1 \alpha) e^{\alpha x_1} + c(x) (e^{\alpha d} - e^{\alpha x_1})\\
  = [-(\alpha^2 a^{1, 1} + b_1 \alpha + c(x) (e^{\alpha (d - x_1)} - 1)] e^{\alpha x_1}
  \leq [-(a^{1, 1} \alpha^2 + b_1 \alpha) + |c(x)| e^{\alpha d}] e^{\alpha x_1} \leq 0
$$
if $\alpha$ is large and $d$ small

Generally we use this to apply the strong maximum principle to small strips of domains
