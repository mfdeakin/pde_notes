## Theorem 1
If $|c(x)|, f \in L^q(B_1), q > \frac{n}{2}$, then
$$
\sup_{B_{1 / 2}} u^+ \leq c ( ||u^+||_{L^2(B_1)} + ||f||_{L^q(B_1)} )
$$
### Proof: Nash-Moser's Proof
Take $v = \eta^2 (\bar{u}_{m}^\beta \bar{u} - k^{\beta + 1})$, $\bar{u} = u^+ + k$,
$$
\bar{u}_m =
\begin{cases}
\bar{u} & u < m\\
m + k & u \geq m
\end{cases}
$$
Then for large $k$
$$
\int |D(w \eta)|^2 \leq C \int |w|^2 (|\nabla \eta|^2 + \eta^2)
$$
By the sobolev embedding, since $w \eta \in H_0^1(B_1)$,

$$
||w \eta||_{L^{\frac{2 n}{n - 2}}(B_1)}^2 \leq C ||\nabla (w \eta)||_{L^2}^2
  \leq C \int_{B_1} |w|^2 (|\nabla|^2 + \eta^2)\\
|\nabla \eta| \leq \frac{C}{R - r}\\
\left( \int_{B_r} w^{2 \frac{n}{n - 2}} \right)^{\frac{n - 2}{n}} \leq \frac{C}{R - r} \int_{B_R} |w|^2
$$
Now we want to say that
$$
m \rightarrow \infty \rightarrow\\
\left( \int_{B_r} |\bar{u}|^{(\beta + 2) \chi} \right)^{\frac{1}{\chi}} \leq \frac{C}{(R - r)^2} \int_{B_R} |\bar{u}|^{\beta + 2}
  \leq \frac{C^{\frac{1}{\beta + 2}}}{(R - r)^{\frac{2}{\beta + 2}}} ||\bar{u}||_{L^{\beta + 2}(B_1)}\\
||u||_{L^{(\beta + 2) \chi}(B_r)} \leq \frac{C^{\frac{1}{\beta + 2}}}{(R - r)^{\frac{2}{\beta + 2}}} ||u||_{L^{\beta + 2}(B_R)}\\
$$
Note we used dominated convergence theorem to take $\int \bar{u}_m^p \rightarrow \int \bar{u}^p$, though it's not valid until we've proven that the bound is bounded.

Next, consider $\beta = 0$, $R = 1$
$$
\beta = 0, R = 1 \rightarrow ||u||_{L^{2 \chi}} \leq C\\
p_0 = 2\\
p_n = \chi p_{n - 1} = 2 \chi^{n - 1}\\
r_n = \frac{1}{2} + \frac{1}{2^{n + 1}}\\
r_{n - 1} - r_n = \frac{1}{2^{n + 1}}\\
a_n = ||u||_{L^{p_n}(B_{r_n})} 
  \leq \frac{C^{\frac{1}{p_{n - 1}}}}{\frac{1}{2^{p_{n - 1} n}}} ||u||_{L^{p_{n - 1}}(r_{n - 1})}
a_n \leq (4^{n + 1} C)^{\frac{1}{2 \chi^{n - 2}}} a_{n - 1}
a_n \leq C^{\frac{n}{\chi^n}} a_{n - 1}
a_n \leq C^{\sum_{j = 1}^n \frac{j}{\chi^j}} a_0
$$
Finally
$$
||\bar{u}||_{L^{p_n}(B_{\frac{1}{2}})} \leq ||\bar{u}||_{L^{p_n}(B_{r_n})}
  \leq C^{\sum_{j = 1}^n \frac{j}{\chi^j}} ||\bar{u}||_{L^2(B_1)}\\
\lim_{n \rightarrow \infty} ||\bar{u}||_{L^{p_n}}(B_\frac{1}{2})
  = ||\bar{u}||_{L^\infty(B_\frac{1}{2})}
  \leq C^{6_0} ||\bar{u}||_{L^2(B_1)}\\
u^+ \leq \bar{u}^+ + k\\
\sup_{B_{\frac{1}{2}}} u^+ \leq C \left( ||u^+||_{L^2(B_1)} + ||f||_{L^q} \right)
$$

De Giorgi's Proof:
Grow an area $u > k$ with linear iterations

$$
\begin{align*}
\int_{\Omega} a^{i, j} \partial_i u \partial_j v + c u v \leq \int_\Omega f v & & \forall v \geq 0\\
I = \int a^{i, j} \partial_i (\partial_j u^2 + (u - k)^+ 2 \eta \partial_j \eta), & & \{u > k\}
\end{align*}
$$

Our Leading Order Term is
$$
\int a^{i, j} \partial_i u \partial_j u \eta^2
  \geq \theta \int |\nabla ((u - k)^+ \eta)|^2
       - \int |a^{i, j}| |\nabla u| (u - k)^+ 2 \eta |\nabla \eta|
  \geq \theta \int |\nabla ((u - k)^+ \eta)|^2 - \int ({(u - k)^+}^2 \eta |\nabla \eta|^2)
$$

$$
\int |f|(u - k)^+ \eta^2 \leq \int \frac{|f|}{k} ((u - k)^+)^2 \eta^2\\
$$

Now we just need to deal with 
$$
\int |c| u (u - k)^+ \eta^2 + |f| (u - k)^+ \eta^2
  \leq \int (|c| + \frac{|f|}{k}) ((u - k)^+ \eta)^2 1
$$
Use Holder's inequality, with the first term as $L^q$, the second as $L^{\frac{2 n}{n - 2}}$, and the 1 as $L^r$,
with
$$
\frac{1}{q} + \frac{1}{\frac{2 n}{n - 2}} + \frac{1}{r} = 1
$$
Then
$$
\int |c| u (u - k)^+ \eta^2 + |f| (u - k)^+ \eta^2
  \leq \int (|c| + \frac{|f|}{k}) ((u - k)^+ \eta)^2 1
  \leq |||c| + \frac{|f|}{k}||_{L^q} ||((u - k)^+ \eta)^2||_{L^\frac{n}{n - 2}}\\

$$
