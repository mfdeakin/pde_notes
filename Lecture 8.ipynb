$$
\begin{cases}

\end{cases}\\
S_g = \{ v | v \text{ is subharmonic in } \Omega, v = g \in \partial \Omega \}
$$

Define $u(x) = \sup_{v \in S_g} v(x)$

Then we will claim that $u$ is the solution. Fix $y_0 \in \Omega$. By definition, $\exists v_n \in S_g$ with $v_n(y_0) \rightarrow u(y_0)$ as $n \rightarrow \infty$.

## Theorem 1

If $g$ is bounded,
$$
S_g = \{ v | v \text{ is subharmonic }, v \leq g \in \partial \Omega \}
$$

Then $u(x) = \sup_{v \in S_g} v(x)$ is harmonic.

Finally we need to show the boundary limit.
That is

$$
\lim_{y \rightarrow x_0 \in \partial \Omega} u(y) = g(x_0)
$$

Provided that $g \in C^0(\partial \Omega)$ and $x_0)$ is a regular point.

### Define
$x_0 \in \partial \Omega$ is called a regular point if $\exists \sup$ harmonic function $w \in \Omega$ s.t. $w > 0$ in $\Omega$ and $w$

$w \in C^0(\bar{\Omega}), w > 0$
Only at $\Omega \setminus \{ x_0 \}$ does the strictly positive work, otherwise $w(x_0) = 0$.

$w$ is called a barrier.

### Proof

$g \in C^0(\partial \Omega) \rightarrow \forall \epsilon > 0, \exists \delta > 0$ s.t. whenever $|x - x_0| < \delta \rightarrow - \epsilon < g(x) - g(x_0) < \epsilon$.
Then $g(x) \leq g(x_0) + \epsilon$, and $k$ s.t. $g(x) \leq k w(x)$.

Then $g(x) \leq g(x_0) + \epsilon + k w(x)$, use this as a barrier.
$g(x_0) + \epsilon + k w(x)$ is a superharmonic function, and $u$ is a subharmonic function, so by definition $u(x) \leq g(x_0) + \epsilon + k w(x)$ whenever $x \in \Omega$, and since $w(x_0) = 0$, we have $\lim_{x \rightarrow x_0, x \in \Omega} u(x) \leq g(x_0) + \epsilon$

Then $g(x_0) - \epsilon - k w \leq g(x)$ on $\partial \Omega$.

Note that our barrier depends on $x_0$.

Can we always find a barrier? Homework!

Barrier is a local property. So if we can find a local barrier, then we're good. Once you can find a barrier at this point, $w \in C^0(N), w > 0$ in $N \setminus \{ x_0 \}$, $w(x_0) = 0$, then we can extend $w$ to be a barrier for the whole space.

Take a small ball and extend it to the minimum outside the ball, and then you can show that this is subharmonic.
