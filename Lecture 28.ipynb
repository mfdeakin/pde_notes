# Part 4, Regularity of Weak Solutions
## Local Estimate ($H^2 = W^{2, 2}$, $H^m = W^{m, 2}$)
### Theorem 1
$a^{i,j} \in C^1$ is Lipschitz, with $B^j, c \in L^\infty$, then if $u$ is a weak solution to
$$
\begin{cases}
L u = f & x \in \Omega\\
u = 0 & x \in \partial \Omega
\end{cases}
$$
where $L u = -\partial_i (a^{i, j} \partial_j u) + b_j \partial_j u + c u$,
then $\forall V \subset \subset \Omega$,
we have $||u||_{W^{2, 2}(V)} \leq C (||u||_{W^{1, 2}(\Omega)} + ||f||_{L^2(\Omega)})$; ie $u \in H^2_{loc}(\Omega)$

#### Proof
Explained Last Time
$$
v = \partial_k^{-h}(\eta^2 \partial_k^h u)\\
\eta = 1 \in \V\\
\sup(\eta) \subset W \subset \subset \Omega\\
\partial_k^h u = \frac{u(x + h e_k) - u(x)}{h}\\
v = \partial_k^{-h}(\eta^2(\partial_k^h u)) \in H_0^1(\Omega)\\
\text{By definition of weak solution}\\
\int a^{i, j} \partial_j u \partial_i v + b_j \partial_j u v + c(x) u v = \int f v\\
\partial_i = \partial_k^{-h}( \partial_i(\eta^2 \partial_k^h u))\\
\int f \partial_k^h g = -\int \partial_k^{-h} f g
$$
Think of this as the weak derivative and integation by parts

Now we want to control all other terms in the leading order term.
Note with this difference operator ($\partial_i^h$) we have the product rule
$$
\int a^{i, j} \partial_j u \partial_i v = -\int \partial_k^h (a^{i, j} \partial_j u) \partial_i (\eta^2 \partial_k^h u)\\
\int a^{i, j} \partial_j u \partial_i v = -\int \partial_k^h(a^{i, j} \partial_j u) \partial_i(\eta^2 \partial_k^h u)\\
= -\int [(\partial_k^h a^{i, j}) \partial_j u + a^{i, j} \partial_k^h \partial_j u]
   \cdot [\partial_i \eta^2 \partial_k^h u + \eta^2 \partial_k^h \partial_i u]
$$
Our leading term is $\int \eta^2 a^{i, j}(\partial_k^h \partial_j u) (\partial_k^h \partial_i u) \geq \theta \int \eta^2 |\partial_k^h D u|^2$

Idea
All other terms can be controlled by $||u||^2_{W^{1, 2}}$ and the leading term.
$\partial_k^h a^{i, j}$ is bounded if it's Lipschitz. $\partial_i \eta^2$ is just bounded
$$
\int \partial_k^h \a^{i, j} + \partial_j u \partial_i(\eta^2) \partial_k^h u \leq \int |D u|^2 + \int \eta^2 |\partial_k^h u|^2\\
\partial_k^h u = \frac{u(x + h e_k) - u(x)}{h} = \frac{1}{h} \int_0^h D_k u(x + s e_k) ds
  = \int_0^1 D_k u(x + t h e_k) dt\\
\int_W |\partial_k^h u|^2 \leq \int_W \int_0^1 |D_k u(x + t h e_k)|^2 dt\\
\leq \text{ (by Fubini) } \int_0^1 \left( \int_W |D_k u(x + t h e_k)|^2 dx \right) dt
\leq \int_0^1 \int_\Omega |D_k u|^2 dx dt = ||D_k u||_{L^2}^2
$$

We took care of one term, now let's take care of the other term
$$
\int \partial_k^h a^{i, j} \partial_j u \eta^2 \partial_k^h \partial_i u \leq \epsilon \int \eta^2 |\partial_k^h \partial_i u|^2 + \frac{|a^{i, j} c^1}{4 \epsilon} \int_\Omega |D u|^2
$$
Now we take care of the cross term of $v$; this doesn't sound difficult but we didn't write anything

Next we consider the $\int b_j \partial_j u \partial_k^{-h}(\eta^2 \partial_k^h u)$ terms
Apply the product rule
$\int b_j \partial_j u ((\partial_k^{-h} \eta^2) \partial_k^h u + \eta^2 \partial_k^{-h} \partial_k^h u)$
We have the first term controlled by the gradient of $u$, so we can ignore it. The second term is harder.
% \partial_k^h u = \frac{u(x + h e_k) - u(x)}{h}\\
% \partial_k^{-h} (\partial^k_h u) = \frac{1}{-h} \left[ \frac{u(x + h e_k - h e_k) - u(x - h e_k)}{h} - \frac{u(x + h e_k) - u(x)}{h} \right]\\
%   = \frac{u(x - h e_k) - u(x + h e_k)}{h^2} 
Just compute the $L^2$ norm of $\int |\partial_k^{-h} (\eta^2 \partial_k^h u)|^2 \leq \int |D (\eta^2 \partial_k^h u)|^2 \leq C \int_\Omega |D(\eta^2) \partial_k^h u|^2 + |\eta^2 D \partial_k^h u|^2
\leq \int_\Omega |D u|^2 + \int \eta^2 |\partial_k^h D u|^2\\
\int_\Omega C u v + \int_\Omega f v \leq \frac{1}{4 \epsilon} \int (u^2 f^2) + \epsilon \int_\Omega v^2
$$
To sum up
$$
\theta \int_\Omega \eta^2 |\partial_k^h D u|^2 \leq \int_\Omega (|D u|^2 + u^2 f^2)
$$

Let's say $w = D u \in L^2$, $\int_V |\partial_k^h w|^2 \leq \int_{\Omega} (|D u|^2 + u^2 + |f|^2)$
#### Claim:
$D_k w$ exists and for any $V \subset \subset W$ $\int_V |D_k w|^2 \leq \int_\Omega (|D u|^2 + u^2 + |f|^2)$

So we have weak convergence, and by Fatou's lemma, this convergence is bounded
$$
\int_W |\partial_k^h w|^2 < \infty \rightarrow \partial_k^h w \rightarrow \text{weak in L^2} w_0,
\int_W |w_0|^2 \leq \lim \int |\partial_k^h w|^2 < \infty\\
\int (\partial_k^h w)\phi = -\int w \partial_k^{-h} \phi
$$
By weak convergence,
$$
\int \partial_k^h w \phi \rightarrow \int w_0 \phi\\
\partial_k^{-h} \rightarrow \partial_k \phi\\
\int w_0 \phi = -\int w \partial_k \phi \rightarrow w_0 = D_k w\\
\int b_j \partial_j u \partial_k^{-h}(\eta^2 \partial_k^h u)\\
\int_{\Omega} v^2 = \int |\partial_k^{-h}(\eta^2 \partial_k^h u)|^2 \leq \int_\Omega |D(\eta^2 \partial_k^h u|^2
\leq \int_\Omega |D(\eta^2) \partial_k^h u|^2 + |\eta^2 D \partial_k^h u|^2\\
\int_V |D_k D u|^2 \leq \int_\Omega (|D \eta|^2 + \eta^2)(|D u|^2 + u^2) + |f|^2
$$

Remark: $v = u$
$$
\int a^{i, j} \partial_j u \partial_i u + \int b_j \partial_j u \u + c u^2 = \int f u\\
\rightarrow \int_{\Omega} |D u|^2 \leq \int_\Omega u^2 + \int_{\Omega} |f|^2\\
||u||_{H^2(V)} \leq C (||u||_{L^2(\Omega)} + ||f||_{L^2(\Omega)})\\
a^{i, j} \in C^1, b_j, c \in L^\infty
$$
Later we can show this for $a^{i, j} \in C^0$ rather than $C^1$

Now we can derive interior regularity, basically if $a^{i, j} \in C^m$, $b_j \in C^{m - 1}$, $c \in C^{m - 1}$, we have $||u||_{H^m(V)} \leq C (||u||_{L^2} + ||f||_{L^2})$

This can be shown by induction and differentitation
$-\partial_k \partial_i(a^{i, j} \partial_j u) \approx -\partial_i(a^{i, j} \partial_j \partial_k u) + lots of terms$

If $u$ is a weak solution, then $\Delta u = f$ in $\Omega$ is a local estimate.
Do we have this global estimate? We'll discuss this next time.
