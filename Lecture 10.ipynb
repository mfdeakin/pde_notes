## Dirichlet Principle

Solve
$$
(1)
\begin{cases}
-\Delta u = f, & \in \Omega\\
u = g, & \in \partial \Omega\\
\end{cases}
f \in C^1(\bar{\Omega}), g \in C^1(\partial \Omega)\\
E[u] = \frac{1}{2} \int_\Omega |\nabla u|^2 - \int_\Omega f u\\
A = \{ u \in C^2(\Omega) \cap C^1(\bar{\Omega}) | u = g \in \partial \Omega \}\\
c = \inf_{u \in A} E[u]\\
$$

1) If $u_0$ satisfies (1), then $u_0 \in A$

2) If $u_0 \in A, E[u_0] = c$, then $u_0$ satisfies (1).

Path to find a solution to (1)

Remarked last time:

1) Not all solutions can be attained by the Dirichlet principle

2) The minimizer may not be attained in $A$, since $A$ is not closed under $\int_\Omega |\nabla|^2$
This can be overcome by looking introducing the Sobolev space, $A \rightarrow \bar{A}$

3) Then we find the solution by the following two processes:
a) Show $c$ is attained in $\bar{A}$, existence part
b) Minimizer is only a weak solution.
c) We need to show that the weak solution is a strong solution, ie that it satisfies the regularity requirements

a) and b) will be done in part 2 of class, c) will be done in part 3

# 2.3 Classical solutions for heat equations

$\Omega = R^n$ is bounded
$u_t = k \Delta u, t > 0, x \in R^n$

Task 1)

$$
\begin{cases}
u_t = \Delta u + f(x, t), & x \in R^n, t > 0\\
u(x, 0) = \phi(x)
\end{cases}
$$

Derive the representation formula formally by the Fourier Transform, proving it rigorously

Use the Fourier transform in $x$, $f(x) \rightarrow \hat{f}(\zeta) = (2 \pi)^{-n/2} \int_{R^n} e^{-i \zeta \cdot x} f(x) dx \rightarrow f(x) = (2 \pi)^{-n/2} \int_{R^n} e^{-i \zeta \cdot x} \hat{f}(\zeta) d\zeta$

$$
u(x, t) \rightarrow \hat{u}(\zeta, t)\\
\Delta u(x, t) \rightarrow -|\zeta|^2 \hat{u}(\zeta, t)\\
f(x, t) \rightarrow \hat{f}(\zeta, t)\\
\phi(x) = \hat{\phi}(\zeta)\\
$$

Then our original problem becomes

$$
\begin{cases}
\hat{u}_t = -|\zeta|^2 \hat{u} + \hat{f}, & \zeta \in R^n, t > 0\\
\hat{u}(\zeta, 0) = \hat{\phi}
\end{cases}\\
\hat{u}(\zeta, t) = \hat{\phi}(\zeta) e^{-|\zeta|^2 t} + \int_0^t e^{-|\zeta|^2 (t - \tau)} \hat{f}(\zeta, \tau)\\
u(x, t) = (2 \pi)^{-n / 2} \int_{R^n} e^{i \zeta \cdot x} \left[ \hat{u}(\zeta, t) \right] d\zeta
$$

First term

$$
(2 \pi)^{-n / 2} \int_{R^n} e^{i \zeta \cdot x - |\zeta|^2 t} \hat{\phi}(\zeta) d\zeta
= (2 \pi)^{-n} \int_{R^n} e^{i \zeta \cdot x - |\zeta|^2 t} \left[ \int_{R^n} e^{-i \zeta \cdot y} \phi(y) dy \right] d\zeta
$$

Fubini's theorem lets us move interchange the integrals, giving

$$
(2 \pi)^{-n} \int_{R^n} \int_{R^n} e^{i \zeta \cdot x - |\zeta|^2 t -i \zeta \cdot y} d\zeta \phi(y) dy\\
= (2 \pi)^{-n} \int_{R^n} \int_{R^n} e^{i \zeta \cdot (x - y) - |\zeta|^2 t} d\zeta \phi(y) dy\\
= (2 \pi)^{-n} \int_{R^n} \int_{R^n} e^{-t \left|\zeta -\frac{i (x - y)}{2 t} \right|^2} d\zeta e^{- \frac{|x - y|^2}{4 t}} \phi(y) dy\\
= (2 \pi)^{-n} \int_{R^n} \int_{R^n} e^{-t |z|^2} dz e^{-\frac{|x - y|^2}{4 t}} \phi(y) dy\\
= (2 \pi)^{-n} t^{-n/2} \int_{R^n} \int_{R^n} e^{-|z|^2} dz e^{-\frac{|x - y|^2}{4 t}} \phi(y) dy\\
= (2 \pi)^{-n} t^{-n/2} \int_{R^n} \pi^{n/2} e^{-\frac{|x - y|^2}{4 t}} \phi(y) dy\\
= (4 \pi t)^{-n / 2} \int_{R^n} e^{-\frac{|x - y|^2}{4 t}} \phi(y) dy\\
$$


Then the second term

$$
(2 \pi)^{-n} \int e^{i \zeta \cdot x} \int_0^t e^{-|\zeta|^2 (t - \tau)} \int e^{-i \zeta \cdot y} f(y, \tau) dy d\tau d\zeta\\
= (2 \pi)^{-n} \int_0^t \int \int e^{i \zeta \cdot (x - y) -|\zeta|^2 (t - \tau)} f(y, \tau) dy d\zeta d\tau\\
= \int_0^t (4 \pi (t - \tau))^{n / 2} \int e^{-\frac{|x - y|^2}{4(t - \tau}} f(y, \tau) dy d\tau\\
$$

Then

$$
\Phi(x, t) = (4 \pi t)^{-n / 2} e^{-\frac{|x|^2}{4 t}}\\
u_t = \Delta u + f(x, t)\\
u(x, 0) = \phi(x)\\
u(x, t) = \int \Phi(x - y, t) \phi(y) dy + \int_0^t \int_{R^n} \Phi(x - y), t - \tau) f(y, \tau) dy d\tau
$$

Properties of $\Phi(x, t)$:

1) $\Phi(x, t) > 0$

2) $\int_{R^n} \Phi(x, t) dx = 1, \forall t > 0$, so it's a probability distribution

3) If $|x| \geq \delta > 0$, then $0 < \Phi(x, t) \leq (4 \pi t)^{-n / 2} e^{-\frac{\delta}{4 t}} \rightarrow 0$ as $t \rightarrow 0$.
Thus, $\Phi(x, t) \rightarrow \delta_0(x)$ as $t \rightarrow 0+$

Step 1

$$
\int \frac{1}{(4 \pi t)^{n / 2}} e^{- \frac{|x - y|^2}{4 t}} \phi(y) dy
$$

Fix $x$ and $t$, then we have $\Phi(x - y, t) \leq c e^{-\frac{|y|^2}{4 t}}$
and $|\phi(y)| \leq e^{a |y|^2} \leq \int e^{-\frac{|y|^2}{4 t} + a |y|^2} dy < \infty$

If $4 t a < 1$, then $t < \frac{1}{4 a}$

In general we need

$\phi(y) \leq e^{\beta(y)}$, with $\frac{\beta(y)}{y^2} \rightarrow 0$ as $|y| \rightarrow \infty$

Then the first integral makes sense, $\forall x \in R^n, t$

There is a similar condition for $f(x, t) \leq e^{\beta(x)}, \frac{\beta(x)}{|x|^2} \rightarrow 0$ as $|x| \rightarrow \infty$

By the end of class, we will see counterexamples

Step 2: Rigorously prove it!

## Theorem 2.3.1

Assume that $|\phi(x)| \leq e^{a |x|^2}$ and $\phi$ is continuous at $x_0$

Then

$$
\begin{cases}
\lim_{t \rightarrow 0+} u(x_0, t) = \phi(x_0)\\
u_t = \Delta u, t > 0
\end{cases}\\
u(x, t) = \int (4 \pi t)^{-n / 2} e^{-\frac{|x - y|^2}{4 t}} \phi(y) dy\\
|u(x, t)| \leq e^{\frac{|x|^2}{4 t}} (4 \pi t)^{-n / 2}\\
u_t = \int \phi_t(x - y, t) \phi(y) dy
$$

Then $u \in C^\infty$ and $\phi_t = \Delta \phi, t > 0$
