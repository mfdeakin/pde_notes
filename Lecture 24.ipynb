3) Characterization of $H^{-1}(\Omega) = (H_0^1(\Omega))^*$

$$
f \in H^{-1}(\Omega) = \{f : H_0^1(\Omega) \rightarrow R | f \text{ bounded, linear}\}
$$

## Theorem 1
1) $f \in H^{-1}(\Omega) \rightarrow \exists f_0, f_1, \ldots, f_n \in L^2$
s.t. $<f, v> = \int_\Omega f_0 v + \sum_{j = 1}^{n} \int_\Omega f_j \partial_j v$.

### Proof
Riesz Representation Theorem
$f: H \rightarrow R$ is bounded, linear, so $<f, v> =  (u, v)_H$, $||f||_{H^{-1}} = ||u||_{H}$

$\exists u \in H_0^1(\Omega)$ s.t. $<f, v> = (u, v)_{H_0^1(\Omega)} = \int_\Omega u v + \int_\Omega D u \cdot D v$,
$||f||_{H^{-1}} = ||u||_{H_0^1(\Omega)}$, $f_0 = u \in L^2$, $f_j = D_j u \in L^2$

2) $||f||_{H^{-1}}(\Omega) = \sup_{v \neq 0}$ $\frac{<f, v>}{||v||_{H_0^1}} = $ $\inf \left\{ \left( \sum_{j = 0}^n \int_\Omega |f_j|^2 dx \right^{1/2} | <f, v> = \int_\Omega f_0 v + \sum_{j = 1}^n \int f_j \partial_j v \right\}$

### Proof
$$
\begin{align*}
<f, v> = &(u, v)_{H_0^1(\Omega)} = \int_\Omega f_0 v + \int_\Omega f_j \partial_j v\\
\leq & \left( \int f_0^2 + \sum \int f_j^2 \right)^{1 / 2} \left( \int v^2 + \sum \int (\partial_j v)^2 \right)^{1 / 2}\\
\leq & \left(\int f_0^2 + \sum \int f_j^2 \right) ||v||_{H_0^1}\\
||f||_{H^{-1}} \leq \sup \frac{<f, v>}{||v||_{H_0^1}} \leq \left( \int f_0^2 + \sum \int f_j^2 \right)^{1 / 2}
\end{align*}
$$

3) $<f, v> = \int_\Omega \left(f_0 - \sum_{j = 1}^n \partial_j f_j\right) v$

### Proof
$$
\bar{C_0^\infty(\Omega)} = H_0^1(\Omega) \rightarrow \int f \partial_j v = -\int \partial_j f_j v
$$
So in the $L^2$ sense,
$$
\begin{align*}
f = &f_0 - \sum \partial_j f_j, f_0 \in L^2, f_j \in L^2\\
H_0^1(\Omega) \subset &L^2(\Omega) \subset H^{-1}(\Omega)\\
\end{align*}
$$

## Definition
A solution $u \in H_0^1(\Omega)$ is called a weak solution of
$$
\begin{cases}
-\Delta u + u = f & \in \Omega\\
u = 0 & \in \partial \Omega
\end{cases}
$$
if
$$
\forall v \in H_0^1(\Omega), \int_\Omega \nabla u \cdot \nabla v + \int_\Omega u v = \int_\Omega f v
$$

# Weak solutions of Elliptic Equations
Note the use of Einstein Summation
$$
L = -\partial_j(a^{i,j} \partial_i u) + \sum_{j = 1}^n b_j \partial_j u + c(x) u\\
a^{i, j}(x) \zeta_i \zeta_j \geq \theta |\zeta|^2, \theta > 0, |a^{i,j}| + |b_j| + |c(x)| \leq C
$$
where $L$ is a uniformly elliptic operator of divergence form.

Extension of the last theorem
Poincare's Inequality 1: $||u||_{L^2(\Omega)}^2 \leq C_0^2 ||\nabla u||_{L^2(\Omega)}^2, u \in H_0^1(\Omega)$

## Theorem 1
$b_j = 0, |c(x)| \leq \frac{1}{2 c_0} \theta$. Then $\forall f \in H^{-1}(\Omega)$, there is a unique weak solution

### Proof
$$
\begin{align*}
<u, v>_{new} = &\int_\Omega a^{i, j} \partial_i u \partial_j v + \int_\Omega c(x) u v\\
  \geq & \theta \int_\Omega |\nabla u|^2 - ||c||_{L^\infty} \int_\Omega u^2\\
  \geq & \theta \int |\nabla u|^2 - \frac{\theta}{2 c_0^2} c_0^2 \int_\Omega |\nabla u|^2\\
  \geq & \frac{\theta}{2} \int_\Omega |\nabla u|^2 \geq c_1 ||u||_{H_0^1}
\end{align*}
$$

$f \in H^{-1}(\Omega) = (H_0^1(\Omega))^* \rightarrow f \in ((H_0^1, < >))^*$
Then Riesz Representation thoerem shows that there is a unique $u \in H_0^1(\Omega)$ s.t. $<u, v>_{new} = <f, v>$, which satisfies the solution of the weak solution.

If $b_j = 0 \rightarrow$ there is a unique weak solution provided $||c||_{L^\infty} \leq \frac{1}{2 c_0^2} \theta$.
What's wrong with $b_j \neq 0$?
Can't switch $\partial_j u v$ and $u \partial_j v$ since it's a first order derivative.
$L$ is not self adjoint; ie $L \neq L^*$
