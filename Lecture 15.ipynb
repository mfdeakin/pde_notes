# Duhumel's Principle
$$
\begin{cases}
u_{tt} = \Delta u + f(x, t)\\
u(x, 0) = 0\\
u_t(x, 0) = 0
\end{cases}\\
u(x, t) = \int_0^t U(x, t; s) ds
$$

$$
\begin{cases}
U_{tt} = \Delta U, t > s\\
U(x, s; s) = 0\\
U_t(x, s; s) = f(x, s)
\end{cases}
$$

For $n = 1$, we have d'Alembert's Formula

$$
\tilde{t} = t - s\\
U(x, t; s) = \frac{1}{2} \int_{x - \tilde{t}}^{x + \tilde{t}} f(y, s) dy\\
  = \frac{1}{2} \int_{x - (t - s)}^{x + (t - s)} f(y, s) dy\\
  = \frac{1}{2} \int_0^t \int_{x - (t - s)}^{x + (t - s)} f(y, s) dy ds
$$

For $n = 3$, we have Kitchhoff's Formula

$$
\begin{cases}
u_{tt} = \Delta u\\
u(x, 0) = g\\
u_t(x, 0) = h
\end{cases}
$$

Apply the same principle as in the $n = 1$ case, we have
$$
\tilde{t} = t - s
U_{\tilde{t}\tilde{t}} = \Delta U\\
\tilde{U} |_{\tilde t = 0} \tilde{U}_{\tilde{t}}|_{\tilde{t} = 0} = f(x, s)\\
U(x, t; s) = \frac{1}{|\partial B_{\tilde{t}}(x)|} \int_{\partial B_{\tilde{t}}(x)} \tilde{t} f(y, s) d\sigma_y\\
u(x, t) = \int_0^t \frac{1}{|\partial B_{\tilde{t}}(x)|} \int_{\partial B_{\tilde{t}}(x)} (t - s) f(y, s) ds\\
  = \int_0^t \frac{1}{|\partial B_{\tilde{t}}(x)|} \int_{\partial B_{\tilde{t}}(x)} \tilde{t} f(y, t - \tilde{t}) d\sigma_y d\tilde{t}\\
  = \int_0^t \frac{\tilde{t}}{|\partial B_{\tilde{t}}(x)|} \int_{\partial B_{\tilde{t}}(x)} f(y, t - \tilde{t}) d\sigma_y d\tilde{t}\\
  = \int_0^t \frac{1}{4 \pi \tilde{t}} \int_{\partial B_{\tilde{t}}(x)} f(y, t - \tilde{t}) d\sigma_y d\tilde{t}\\
  = \int_0^t \frac{1}{4 \pi} \int_{\partial B_{t}(x)} \frac{f(y, t - |y - x|)}{|y - x|} d\sigma_y d\tilde{t}
$$

$n = 2$, $n \geq 4$ can be done similarly

# Summary

$$
\begin{cases}
u_{tt} = \Delta u + f(x, t)\\
u(x, 0) = g(x)\\
u_t(x, 0) = h(x)
\end{cases}
$$

$n = 1$
$$
u = \frac{1}{2} [g(x + t) + g(x - t)] + \frac{1}{2} \int_{x - t}^{x + t} h(y) dy + \frac{1}{2} \int_0^t \int_{x - (t - s)}^{x + (t - s)} f(y, s) dy ds
$$

$n = 3$
$$
u = \frac{1}{|\partial B_t(x)|} \int_{\partial B_t(x)} [t h + g + \nabla g \cdot (y - x)] d\sigma_y + \frac{1}{4 \pi} \int_{B_t(x)} \frac{f(y, t - |y - x|)}{|y - x|} dy
$$

# Consequences

If $g = h = 0$ in $|x| > R$, the $u = 0$ in $|x| > R + t$
We can prove it with the energy method.
$$
E(t) = \frac{1}{2} \int_{\Omega} (u_t^2 + |\nabla u|^2) dx
$$

## Theorem 1 (Finite Propogation)

If $u(x, 0) = u_t(x, 0) = 0$ in $B_{t_0}(x_0)$,
the cone with vertex at $x_0, 0$ with axis as $t$ increases and radius $r = t$,
then $K(x_0, t_0) = \{ (x, t) | |x - x_0| \leq t_0 - t \} = \cup_{0 \leq t \leq t_0} \bar{B}_{t}(x_0)$

Prove that outside this cone, $u = 0$, either by the formula or with the energy method

### Proof
Let $\Omega = B_{t_0 - t}(x_0)$
$$
E(t) = \frac{1}{2} \int_{B_{t_0 - t}(x_0)} (u_t^2 + |\nabla u|^2\\
\frac{d E}{dt} = \int_{B_{t_0 - t}(x_0)} (u_{tt} + \nabla u \cdot \nabla u_t d\sigma_y - \frac{1}{2} \int_{\partial B_{t_0 - t}(x_0)} (u_t^2 + |\nabla u|^2) d\sigma_y\\
$$

$$
\int_{B_{t_0 - t}(x_0)} \nabla u \cdot \nabla u_t = \int_{B_{t_0 - t}(x_0)} [\nabla(u_t \nabla u) - u_t \Delta u] d\sigma_y
  = \int_{B_{t_0 - t}(x_0)} u_t \frac{\partial u}{\partial n} - \int_{B_{t_0 - t}(x_0)} u_t \Delta u
$$

Then

$$
\frac{d E}{dt} = \int_{B_{t_0 - t}(x_0)} u_t [u_{tt} - \Delta u] + \int_{\partial B_{t_0 - t}(x)} u_t \frac{\partial u}{\partial \nu} dy - \frac{1}{2} \int_{\partial B_{t_0 - t}(x_0)} (u_t^2 + |\nabla u|^2) d\sigma_y)\\
  \leq 0
$$
for $0 < t < t_0$

Then $E(0) = \frac{1}{2} \int_{B_{t_0}(x_0)} (u_t^2 + |\nabla u|^2) = 0$

$0 \leq E(t) \leq 0 \rightarrow u_t = \nabla u = 0$ in $B_{t_0 - t}(x_0) \forall t$

How do we finish our proof?

Simple picture...

## Corollary 1
If $u = u_t = 0$ in $|x| > R$, $u_{tt} = \Delta u$, then $u = 0$ in $|x| > R + t$

## Corollary 2

$$
u_{tt} = \Delta u\\
u(x, 0) = 0, u_t(x, 0) = 0
$$

has a unique solution.

# Energy for bounded domains

Have a bounded domain $\Omega$, and look at the parabolic cylinder and the wave equation, $u(x, 0) = g(x)$ as the initial condition, and Dirichlet boundary conditions on the domains boundary $u = f$.
Use the same proof
Claim that there's a unique solution to the mixed value problem
$$
u_{tt} = \Delta u, x \in \Omega, 0 < t < T\\
u(x, 0) = g(x)\\
u_t(x, 0) = h(x)
u(x, t) = f(x, t), x \in \partial \Omega, 0 < t < T\\
$$

$$
E = \int_{\Omega} \frac{1}{2} [u_t^2 + |\nabla u|^2] dx\\
\frac{dE}{dt} = \int_{\Omega} [u_t u_{tt} + \nabla u \cdot \nabla u_t] dx
$$

We've discussed the three second order equations and their strong solutions
$\Delta u = f$, representation formula
$\Delta u = f$ in a bounded domain, Green's representation formula and Perron's Method proving the existence of the Green's function
Heat equation, $u_t = \Delta u + f$
Wave equation, $u_{tt} = \Delta u + f$
These equations hold pointwisely, ie their limits approach the boundaries

This is the classical theory in PDEs, up the the end of the 50s.
Next we discuss the modern theories, started in the 60s with Sobolev spaces!!!

Strong solutions are replaced by weak solutions, a whole theory on the regularity of weak solutions.
Regularity of weak solutions: 70's Schauder theory, $H^2$ theory, $W^{2, p}$ theory, 80's Evans-Krylov theory
