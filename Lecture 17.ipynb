### Proof

$u^+ = f^+(u)$,
$$
f^+ =
\begin{cases}
x & x > 0\\
0 & x \leq 0
\end{cases}
$$

$$
f_\epsilon^+ =
\begin{cases}
\sqrt{x^2 + \epsilon^2} - \epsilon & x \geq 0\\
0 & x < 0
\end{cases}
$$

By our corollary
$$
D f_\epsilon^+(u) = f_\epsilon'(u) Du \leftarrow \rightarrow \int f'(u) Du \phi = -\int f^+_\epsilon(u) D\phi
$$
Can take $\epsilon \rightarrow 0$ because of the dominated convergence theorem, since $f^+_\epsilon$ is bounded by $1$.
Then
$$
\lim_{\epsilon \rightarrow 0} \int_{u > 0} Du \phi = - \int_{Du^+ = Du X_{u > 0}} u^+ D\phi
$$

And therefore
$$
D|u| = Du^+ - Du^- =
\begin{cases}
D u & u > 0\\
-D u & u < 0\\
\end{cases}\\
|D |u|| = |D u|\\
u, v \in W^{1, p}(\Omega) \rightarrow \max(u, v), \min(u, v) \in W^{1, p}\\
\max(u, v) = \frac{1}{2}(u + v + |u - v|)
$$

Now we can prove the density theorem, $C^\infty(\Omega)$ is dense in $W^{k, p}(\Omega)$, assuming that $\Omega$ is a bounded domain.
$\Omega_j = \{ x \in \Omega | d(x, \partial \Omega) > \frac{1}{j} \}, \Omega = \cup_{j = 1}^\infty \Omega_j$

$\exists$ a partition of unit, subcoordinate to $\{ \Omega_j \}$,
$$\eta_j \in C_0^\infty(\Omega),\\
\eta_j =
\begin{cases}
1 & x \in \Omega_{j + 1} \setminus \Omega_j\\
0 & x \in \Omega_{j - 1}, \Omega \setminus \Omega_{j + 2}
\end{cases}\\
\sum_{j = 1}^\infty \eta_j = 1\\
u = \left( \sum \eta_j \right) u = \sum (\eta_j u)
$$

Fix $\delta > 0$. For each $j$, $\eta_j u$ has compact support.
Then $\exists \epsilon_j$ s.t. $||u_{\epsilon_j} - \eta_j u||_{W^{k,p}(\Omega)} < \frac{\delta}{2^{j + 1}}$, the support is one block from this one.
The support of $u_{\epsilon_j}$ is a subset of $\Omega_{j + 3} \setminus \Omega_{j - 2}$

$$u
u - \sum_{j = 1}^{\infty} u_{\epsilon_j} = \sum_{j = 1}^{\infty} (\eta_j u -  u_{\epsilon_j})\\
||u - \sum_{j = 1}^{\infty} u_{\epsilon_j}||_{W^{k, p}(\Omega)} = ||\sum_{j = 1}^{\infty} (\eta_j u -  u_{\epsilon_j})||_{W^{k, p}(\Omega)} \leq \sum_{j = 1}^{\infty} ||\eta_j u -  u_{\epsilon_j}||_{W^{k, p}(\Omega)}

$$

## Extension Property
If $u \in W^{k, p}(\Omega)$, then $\exists$ a linear bounded extension operator $E: W^{k, p}(\Omega) \rightarrow W^{k, p}(R^n)$ s.t.
1) $E u = u$ almost everywhere in $\Omega$

2) You have control of your extent, $||E u||_{W^{k, p}(R^n)} \leq C ||u||_{W^{k, p}(\Omega)$ for some $C$ dependent on the domain and the dimension

### Proof
$\Omega = B^+_1 = B_1(0) \cap R_+^n = B_1(0) \cap \{ x_n > 0 \}$, with $u \in C^{\infty}(\bar{B}^+_1)$
First consider the case where $k = 1$. The boundary is flat, and we just want to extend beneath the axis s.t. $u(x', x_n)$ is continuous about $x_n = 0$
The right extension is
$$
\tilde{u}(x', x_n) =
\begin{cases}
u(x', x_n), & x_n \geq 0\\
u(x', -x_n), & x_n < 0
\end{cases}
$$
This extension is continuous, and
$$
D_{x_n} \tilde{u}(x', x_n) =
\begin{cases}
D_{x_n} u(x', x_n) & x_n \geq 0\\
-D u_{x_n}(x', -x_n) & x_n < 0
\end{cases}\\
||\tilde{u}||_{L^p(B_1)} + ||D \tilde{u}||_{L^p(B_1)} \leq 2 ||u||_{L^p} + 2 ||D u||_{L^1(B^+_1)}
