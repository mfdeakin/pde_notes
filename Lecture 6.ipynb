### Proof of 4:

$a \neq b, a, b \in \Omega$

$u(x) = G(x, a), v(x) = G(x, b)$
Then take two balls $B_\epsilon(a), B_\epsilon(b)$ around $a, b$ each exclusive

Then $\Delta u = \Delta v = 0$ in $S = \Omega \setminus \left(B_\epsilon(a) \cup B_\epsilon(b)\right)

Then $\int_{S} (u \Delta v - v \Delta u) = \int_{\partial \Omega} (u \frac{\partial v}{\partial \nu} - v \frac{\partial u}{\partial \nu} ) - \int_{\partial B_\epsilon(a)} \left(u \frac{\partial v}{\partial \nu} - v \frac{\partial u}{\partial \nu} \right) - \int_{\partial B_\epsilon(b)} \left(u \frac{\partial v}{\partial \nu} - v \frac{\partial u}{\partial \nu} \right)$


$u \Delta v = v \Delta u = 0$ on $S$ since these are Harmonic functions.

Also $u \frac{\partial v}{\partial \nu} = v \frac{\partial u}{\partial \nu} = 0$ on $\partial \Omega$

So we want to compute the remaining two integrals.

$\int_{\partial B_{\epsilon}(a)} u \frac{\partial v}{\partial \nu} = 0$, since on the boundaries of the balls, $u = G(x, a) = \Phi(|x - a|) - \Phi^x(a) \approx \Phi(|x - a|)$ Then by the same proof in lecture 1, it goes to 0.

Also, $\int_{B_\epsilon(a)} v \frac{\partial u}{\partial \nu} \rightarrow - v(a)$, $\int_{B_\epsilon(b)} u \frac{\partial v}{\partial \nu} \rightarrow u(b)$, $\int_{B_\epsilon(b)} v \frac{\partial u}{\partial \nu} \rightarrow 0$

Therefore, $u(b) = v(a) \rightarrow G(b, a) = G(a, b)$
