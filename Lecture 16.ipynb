# Theorem 2: The chain rule for weak derivatives
$u \in L_{loc}^1, D u$ exists. Suppose $f \in C^1(R)$ and $f' \in L^\infty$. Then $D(f(u)) = f'(u) Du$

## Proof

We don't know anything about $f(u)$ Use that $u_\epsilon \in C^\infty(\Omega), u_m \rightarrow u$ in $L_{loc}^1, D u_m \rightarrow u$ in $L_{loc}^1$
$D f(u_m)$ is a $C^1$ function
So by the usual chain rule for the definition of derivatives, we have, $D f(u_m) = f'(u_m) D u_m$
Then $\int D(f(u_m)) \phi = -\int f(u_m) D \phi = \int f'(u_m) D u_m \phi$

Then $\int f'(u_m) Du \phi$ is our result, as desired.
