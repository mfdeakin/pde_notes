Properties of $W^{k, p}(\Omega)$

1) Extension
# Theorem 1
Suppose $\bar{\Omega} \subset\subset V$. Then $\exists E: W^{1, p}(\Omega) \rightarrow W^{1, p}(V)$, an extension operator, s.t.

1) $Eu = u$ in $\Omega$

2) $||Eu||_{W^{1, p}(V)} \leq C ||u||_{W^{1, p}(\Omega)}$

## Proof
By the density theorem, we can just consider $u \in C^\infty(\bar{\Omega})$.

### Step 1 (Simplest Case): $\Omega = B^+_{r_0}$

$$
Eu(x', x_n) =
\begin{cases}
u(x', x_n) & x_n \geq 0\\
u(x', -x_n) & x_n < 0
\end{cases}
$$

$Eu \in C^0, ||Eu||_{W^{1, p}(B_{r_0})} \leq 2 ||u||_{W^{1, p}(B^+_{r_0})}$

### Step 2
$x_0 \in \partial \Omega, \exists B_{r_{x_0}}(x_0)$ and
$\Phi_{x_0}$ s.t. $B_{r_{x_0}}(x_0) \cap \Omega \rightarrow B^+_1$ and $\Psi_{x_0} = (\Phi_{x_0})^{-1}$
$\bar{\Omega}^{compact} \subset \cup_{x_0 \in \bar{\Omega}} B_{r_{x_0}}(x_0) \subset V$

$\bar{\Omega}^{cpt} \subset \cup_{x_j \in \bar{\Omega}}^m B_{r_{x_j}}(x_j) = \cup_{x_j \in \Omega} B_{r_{x_j}}(x_j) \cup \cup_{x_j \in \partial \Omega} B_{r_{x_j}}(x_j)$
This is a partition of unit, st. $\sum_{j=1}^{m} \eta_j = 1$
$Eu = \sum_{x_j \in \Omega} (\eta, u) + \sum_{x_j \in \partial \Omega} E^\#[\eta_j u(\psi_{x_j}(y))] \cdot \Phi_{x_j}$
The $\Psi$ takes $B_1 \rightarrow \Omega$, and $\Phi$ takes $\Omega$ to $B_1$.
Then
$||Eu||_{W^{1, p}(V)} \leq C||u||_{W^{1, p}(\Omega)}$

### Step 3: Density Argument
$u \in W^{1, p}(\Omega) \rightarrow \exists u_m \in C^{\infty}(\Omega)$\\
$||u_m - u|_{W^{1, p}} \rightarrow 0$\\
$||u_m - u_k||_{W^{1, p}} \rightarrow 0$, $m, k \rightarrow \infty$\\
$E[u_m] = u_m$ in $\Omega$\\
$||E[u_m] - E[u_k]||_{W^{1, p}(V)} \leq C ||u_m - u_k||_{W^{1, p}(\Omega)} \rightarrow 0$\\
Then $E[u_m]$ is a Cauchy sequence in $W^{1, p}(V)$,
and $E[u] = \lim_{m \rightarrow \infty} E[u_m] \in W^{1, p}(V)$,
$||E[u]||_{W^{1, p}(V)} \leq C ||u||_{W^{1, p}(\Omega)$

Uniform lipshitz bound at infinity then the same thing works for unbounded domains

## Remark
$W^{k, p}$, $k \geq 2$
$$
E[u] =
\begin{cases}
u(x', x_n), & x_n \geq 0\\
\sum_{j = 1}^{k} c_j u(x', -x_n / j), & x_n < 0
\end{cases}\\
E[u] \in C^{k - 1}\\
\sum_{j=1}^k c_j \left( -\frac{1}{j} \right)^l = 1, l = 0, k - 1\\
k = 1, c_1 = 1\\
k = 2, c_1 = 3, c_2 = -4\\
k = 4, \text{ Homework }
$$

Standard argument
Assume $C^\infty$ and work from there

$D_{x_n} u(x', 0+) = D_{x_n} u(x', 0-) \leftarrow \rightarrow \sum_{j = 1}^k c_j \left( c_j \frac{1}{j} \right) = 1$

# 2. Trace

## Theorem 2: $\partial \Omega \in C^1$
$\exists$ a linear operator
$T: W^{1, p}(\Omega) \rightarrow L^p(\partial \Omega)$
s.t.
$||Tu||_{L^p(\partial \Omega)} \leq C||u||_{W^{1, p}(\Omega)}$
Tricks to define this trace operator, we need a weak derivative

### Proof
#### Step 1, Simple case $B^+_1$, with $u = 0$ on $\partial B_1$
Want to define something on this trace
What happens is we take $u(x', 0)$, raise it to the power $p$ (can do since it's $C^\infty$)
$$
|u|^p(x', 0) = -\int_0^1 \frac{\partial}{\partial x_n} (|u|^p(x', x_n)) dx_n
= -\int_0^1 \frac{\partial}{\partial x_n} (|u|^p(x', 1) - |u|^p(x', \something)) dx_n
= \int |u|^p(x', 0) dx' = -\int \left( \int_0^1 p |u|^{p - 1} \frac{\partial}{\partial x_n} |u|(x', x_n) dx_n \right) dx'
$$
This is nothing but the double derivative, $|u|(x', x_n) dx_n$

We can take the gradient
$$
\int_{\partial B^+_1} |u|^p \leq p \int_{B^+_1} |u|^{p - 1} |\nabla u| dx\\
  \leq c \int_{B^+_1} |u|^p + c \int_{B^+_1} |D u|^p\\
||u||_{L^p(\partial B^+_1)} \leq C ||u||_{W^{1, p}(B^+_1)}
$$
In the first term of the last equation, the trace is well defined.

#### Step 2
Take the boundary of the domain, cover it in balls, then map
$$
\partial \Omega \subset \cup_{x_j \in \partial \Omega} B_{r_j}(x_j)
$$
Then the same argument as earlier
$Tu = \sum T[\eta_j u \cdot \Psi_j] \cdot \Phi_j$, and then convergence of the Cauchy sequence

## Corollary 1
$u \in W^{1, p}_0(\Omega)$
Then $Tu = 0$ on $\partial \Omega$

### Proof
Follows from the definition
Because $u_m$ is $C_0^\infty(\Omega)$, and $Tu_m = 0$
So we take the limit of this, and $||T u_m - Tu||_{L^p} \leq C ||u_m - u||_{W^{1, p}}$ which converges to $0$

## Corollary 2
If $u \in W^{1, p}(\Omega)$ and $Tu = 0$, then $u \in W^{1, p}_0(\Omega)$

### Proof
Simple case first, $R^n_+$
$u \in W^{1, p}(R^n_+) \rightarrow \exists u_m \in W^{1, p}(R_+^n), u_m \in C^\infty, u_m \rightarrow u$
$Tu = 0$
$T u_m(x', 0) = u_m(x', 0) \rightarrow 0$ in $L^p(\partial R^n_+)$
Question: can we produce $\tilde{u}_m \in C^\infty_0(R^n_+)$ s.t. $\tilde{u}_m \rightarrow u$ in $W^{1, p}(R^n_+)$?

Cut a $u$, take the cut function $u_\epsilon = u \left(1 - \eta \left(\frac{1}{\epsilon} \right) \right) = 0$ on the boundadry $\partial R^n_+$, where
$$
\eta =
\begin{cases}
0 & t < 1\\
1 & t > 2
\end{cases}\\
$$
Claim
$||u_\epsilon - u||_{W^{1, p}} \rightarrow u \in W^{1, p}_0$
