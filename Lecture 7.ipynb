Aim: Prove the existence of a solution to
$$
\begin{cases}
\Delta u(x) = 0, & x \in \Omega\\
u(x) = 0, & x \in \partial \Omega\\
\end{cases}
$$

## Theorem: Perron's Method
If $\Omega \in C^2, g \in C^0(\partial \Omega)$, then there exists a solution.

First simple case: $\Omega = B_R$

### Theorem 1
$\Omega = B_R, g \in C^0(\partial B_R)$, then let
$$
u(x) = \int_{\partial B_R} \frac{R^2 - |x|^2}{\omega_n R |x - y|^n} g(y) d\sigma_y, x \in B_r
$$
Let $K(x, y) = \frac{R^2 - |x|^2}{\omega_n R |x - y|^n}$.

Then
$\Delta u(x) = 0$, as $u(x) = \int_\partial g(y) \frac{\partial}{\partial \nu} \left( \Phi(x - y) - \phi^x(y) \right) d\sigma_y$.

Need to prove that $u(x_0) = g(x_0)$ for $x_0 \in \partial \Omega$ iff $\lim_{x \rightarrow x_0, x \in B_r} u(x) = g(x_0)$

Note $u(x) - g(x_0) = \int_{\partial B_R K(x, y) g(y) d\sigma_y - g(x_0) \int_{\partial B_R} K(x, y) dy$

Fix $\epsilon > 0$, so $g \in C^0(\partial B_R) \rightarrow \exists \delta > 0, |y - x_0| < \delta$ s.t. $|g(y) - g(x_0)| < \epsilon$

Then $\int_{\partial B_R \cap B_\delta(x_0)} K(x, y) \left( g(y) - g(x_0) \right) d\sigma_y + \int_{\partial B_R, |y - x_0| > \delta} K(x, y) (g(y) - g(x_0)) d\sigma_y = I_1 + I_2$

Then

$$
|I_1| \leq \int_{\partial B_R \cap B_{\delta}(x_0)} K(x, y) \epsilon d\sigma_y \leq \epsilon\\
|I_2| \leq 2 \max_{\partial B_R} |g| \int_{|y - x_0| > \delta, |y| = R} \frac{R^2 - |x|^2}{\omega_n R |x - y|^n} d\sigma_y
$$

Perron's method by sub-super-solution

## Definition 1
$u$ is a subharmonic function in $\Omega$ if $u \in C^0(\Omega) \forall B_R(x_0) \subset \Omega, B_R(x_0) \neq \Omega, \Delta h(x) = 0$ for $x \in B_R(x_0), h(x) \geq u(x)$ for $x \in \partial B_R(x_0)$.

Subharmonic functions satisfy $u \leq h$ in $B_R(x_0)$

### Lemma 1
$u$ is subharmonic, then $u$ satisfies the "sub-M-V-P", ie $u(x_0) \leq \frac{1}{|\partial B_R(x_0)|} \int_{\partial B_R(x_0)} u$

#### Proof
1) Solve
$$
\begin{cases}
\Delta h = 0, & x \in B_R(x_0)\\
h = u, & x \in \partial B_R(x_0)
\end{cases}
$$
By Theorem 1, $h$ exists

2) By definition, $u(x_0) \leq h(x_0) = \frac{1}{|\partial B_R(x_0)|} \int_{\partial B_R(x_0)} u d\sigma_y$

### Lemma 2

$u$ subharmonic if $\max_{\bar{\Omega}} u = \max_{\partial \Omega} u$

$u$ can't attain a maximum inside $\Omega$ unless $u$ is constant.

#### Proof

"Same proof" as for Harmonic functions

### Corollary

$u \in C^0(\Omega)$, subharmonic iff $u$ satisfies the sub-harmonic M-V-P.

#### Proof

$\rightarrow$: Lemma 1

$\leftarrow$: $u$ satisfies the sub-M-V-P, so we can find $h$, and then $u - h$ satisfies the sub-M-V-P in $B_R(x_0)$,
so $\max_{\bar{B_R(x_0)}} (u - h) = \max_{\partial B_R(x_0)} (u - h) \leq 0$

## Proof of some method

### Corollary
If $u$ is subharmonic, $v$ is superharmonic, $u(x) \leq v(x)$ for $x \in \partial \Omega$, then $u(x) \leq v(x)$ for $x \in \Omega$

#### Proof
$u - v$ is easily seen to be subharmonic, so by the MVP, $u \leq v$ on $\partial \Omega$

### Theorem 1
If $\Omega \subset C^2, g \in C^0(\partial \bar{\Omega})$, then there is a solution to
$$
\begin{cases}
\Delta u = 0, & x \in \Omega\\
u = g, & x \in \partial \Omega
\end{cases}
$$

#### Proof

$S_g = \left\{ u \in C^0(\Omega), \text{ subharmonic in } \Omega, u \leq g \text{ for } x \in \partial \Omega \right\}$

### Theorem 2

The function

$u(x) = \sup_{v \in S_g} v(x)$, $u$ is harmonic in $\Omega$.

#### Proof

1) $S_g \neq \emptyset$
Just take $u(x) = \min_{y \in \partial \Omega} g(y) \in S_g$

2) $S_g$ is bounded from above.
$\max_{\partial \Omega} g$ is a harmonic function.

$\forall u \in S_g, u \leq g$ on $\partial \Omega$, then $u \leq \max_{\partial \Omega} g$. Then by the M.P., $u \leq \max_{\partial \Omega} g$

By 1) and 2), we know $u(x) = \sup_{v \in S_g} v(x)$ is well defined. Then $\min_{\partial \Omega} g \leq u(x) \leq \max_{\partial \Omega} g$

3) Let $v \in S_g$ and $\bar{v}$ be the solution
$$
\begin{cases}
\Delta \bar{v}(x) = 0 & x \in B_R(x_0)\\
\bar{v}(x) = v & x \in \partial B_R(x_0)
\end{cases}
$$

Then let
$$
V(x) =
\begin{cases}
\bar{v}(x), & x \in B_R(x_0)\\
v(x), & x \in \Omega \setminus B_R(x_0)
\end{cases}
$$

We call $V$ the harmonic lifting of $v$.

Then $V \in S_g, v(x) \leq V(x)$ for $x \in \Omega$

##### Now we just need to show that $V$ is subharmonic, then $v$ will also be subharmonic.

Let $h$ satisfy
$$
\begin{cases}
\Delta h(x) = 0, & x \in B_R(x'_0) = B'\\
h = V, & x \in \partial B'
\end{cases}
$$

If $B'$ is inside the original $B_R$, then we're fine, since that's satisfied, and if $B_R \cap B' = \emptyset$, we're fine. The issue is when $B_R \cap B' \neq \emptyset$. Consider this case.

On the intersection part, we have two harmonic functions. We can use the maximum principle, then show that $h \geq v$
