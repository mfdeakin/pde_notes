3) Maximum principle

### Theorem 2.2.8
$-\Delta u = 0\ \text{in}\ \Omega$, then $\max_{\Omega \cup \partial \Omega} u = \max_{\partial \Omega} u$
and $u$ cannot attain it's maximum in the domain unless $u$ is constant.
The last requirement of not attaining its maximum in the domain is the strong maximum principle.

#### Proof
Let $\max_{\Omega \cup \partial \Omega} u(x) = u(p_0)$

If $p_0 \in \partial \Omega$, then we're done.

Suppose $p_0 \in \Omega$.

Draw a small ball $B_r$ around $p_0$ in the domain.

Then by the mean value principle, $u(p_0) = \frac{1}{|B_r|} \int_{B_r(p_0)} u d\sigma \leq \frac{1}{|B_r|} \int_{B_r} u(p_0) = u(p_0)$.

Thus, $u$ must be constant in the ball $B_r$.

We can do cover any curve in $\Omega$ by small balls in $\Omega$ which require that $u(x) = u(p_0)$ which is constant.

QED

### Corollary 1
If $\exists$ a solution to $-\Delta u = f(x), x \in \Omega, u = g, x \in \partial \Omega$, then the solution is unique.

#### Proof
Let $u_1, u_2$ be solutions to the previous equation. Then $-\Delta (u_1 - u_2) = 0$ by linearity, so $\max_{p \in \Omega} u_1 - u_2 = u_1(p_0) - u_2(p_0)$, and either $p_0 \in \partial \Omega$, or $u_1 - u_2$ is constant. On the boundary, $u_1 = u_2 = g$, so the maximum on the boundary is $0$, and thus the maximum is $0$. This also holds for $u_2 - u_1$, so they are identical.
