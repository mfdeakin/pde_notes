## Theorem: Weak Maximum Principle
We want to apply the principle to solutions of the PDE rather than just supersolutions
$L u \geq 0$ in $\Omega$ implies $\max_{\bar{\Omega}} u \leq \max_{\partial \Omega} u^+$

### Proof
A standard trick: Add an auxiliary function
$$
u(x) < w(x) = u(x) + \epsilon e^{\alpha x_1}\\
L w = L u + \epsilon e^{\alpha x_1} (\alpha^2 a_{1,1} + \alpha b_1 + c)\\
$$
Since $a_{1, 1} \geq \theta$, $b_1$ and $c$ are bounded, we can choose $\alpha$ sufficiently large that $L w > 0$
$$
\epsilon \downto 0 \rightarrow \max_{\bar{\Omega}} u \leq \max{u} \leq \max_{\partial \Omega} u^+
$$

### Theorem: Hopf Lemma
On an open ball $B \subset R^n$, we have
$$
x_0 \in \partial B\\
u \in C^2(B) \cap C(B \cup \{x_0\})\
c \leq 0\\
\begin{cases}
L u \geq 0 & \in B
\end{cases}\\
$$
Suppose $u(x) < u(x_0) \forall x \in B, u(x_0) \geq 0$
Then looking at any direction $d$ going out from the boundary of the ball (ie, $d \cdot \nu > 0$)
we have
$$
\liminf_{t \rightarrow 0+} \frac{u(x_0) - u(x_0 - t d)}{t} > 0
$$
