# Wave Equation Solutions
$$
\begin{cases}
u_{tt} = \Delta u\\
u(x, 0) = g(x)\\
u_t(x, 0) = h(x)
\end{cases}
$$

### d'Alembert's Formula $n = 1$
$u = \frac{1}{2} [g(x + t) + g(x - t)] + \frac{1}{2 c} \int \limits_{x-t}^{x+t} h(y) dy$

### Kitchhoff's Formula $n = 3$
$u(x, t) = \frac{1}{|\partial B_t(x)|} \int_{\partial B_t(x)} [t h(y) + g(y) + \nabla g(y) (y - x)] d\sigma_y$

### Hadamard's Method of Descent $n = 2$

$$
\tilde{u}(x_1, x_2, x_3, t) = u(x_1, x_2, t)\\
\tilde{g}(x_1, x_2, x_3) = g(x_1, x_2)\\
\tilde{h}(x_1, x_2, x_3) = h(x_1, x_2)\\
\tilde{u}(x_1, x_2, x_3, t) = u(x_1, x_2, t)
= \frac{1}{|\partial B_t((x_1, x_2, 0))|} \int_{\partial B_t((x_1, x_2, 0))} [t h(y_1, y_2) + \tilde{g}(y_1, y_2) + \nabla \tilde{g}(y_1, y_2)(y - x)] d\sigma_y\\
$$

$$
\partial B_t((x, 0)) = \{ |y_1 - x_1|^2 + |y_2 - x_2|^2 + z^2 = t^2\} = \{ z = \sqrt{t^2 - |y - x|^2} \}
\int_{\partial B_t((x, 0))} h(y_1, y_2) d\sigma_y
= 2 \int_{\{ z = \sqrt{t^2 - |y - x|^2} \}} h(y) d \sigma_y
= 2 \int \int_{|y - x| \leq t} h(y) \sqrt{1 + |\nabla \eta|^2} dx dy
= 2 \int \int_{|y - x| \leq t} \frac{h(y)}{\sqrt{t^2 - |y - x|^2}} dy
$$
since
$$
\eta(y) = \sqrt{t^2 - |y - x|^2} \rightarrow |\nabla \eta|^2 = \frac{|y - x|^2}{t^2 - |y - x|^2}
$$

Next
$$
\frac{1}{|\partial B_t|} \int_{\partial B_t} g + \nabla g(y - x) d\sigma_y
= \frac{\partial}{\partial t} [ t \averageint_{\partial B_t} g]
= \frac{\partial}{\partial t} \left[ \frac{z t^2}{|\partial B_t(x)|} \int \int_{|y - x| \leq t} \frac{g}{\sqrt{t^2 - |x - y|^2}} dy \right]
$$

$$
\int \int_{|y - x| \leq t} \frac{g(y)}{\sqrt{t^2 - |y - x|^2}} dy
= \int \int_{y = x + t z, |z| \leq 1} \frac{g(x + t z)}{t \sqrt{1 - |z|^2}} dz
= t \int \int_{|z| \leq 1} \frac{g(x + t z)}{\sqrt{1 - |z|^2}} dz
$$

$$
\frac{\partial}{\partial t} \left[ \int \int_{|y - x| \leq t} \frac{g(y)}{\sqrt{t^2 - |y - x|^2}} dy \right]
= \frac{\partial}{\partial t} \left[ t \int \int_{|z| \leq 1} \frac{g(x + t z)}{\sqrt{1 - |z|^2}} dz \right]
= \int \int_{|z| \leq 1} \frac{g(x + t z)}{\sqrt{1 - |z|^2}} dz + t \int \int_{|z| \leq 1} \frac{\nabla g \cdot z}{\sqrt{1 - |z|^2}} dz
$$

Thus, for $n = 2$, we have

$$
u(x, t) = \frac{1}{2} \averageint_{B_t(x)} \frac{t^2 h + t g + t \nabla g \cdot (y - x)}{\sqrt{t^2 - |y - x|^2}} dy
$$

What about higher dimensions? $n = 5$, we have a trick

$n = 5$

Averaging to reduce the dimension?

$$
U_{tt} = U_{rr} + \frac{4}{r} U_r\\
U = r^{-3} \tilde{U}\\
\tilde{U} = r U \rightarrow U = r^{-1} \tilde{U}\\
\Delta \left( \frac{1}{r^{n - 2} \right) = 0
$$

Note that
$$
\Delta \left( \frac{1}{r^{n - 2}} \right) \neq 0
$$

Then

$$
\Delta U = \Delta (r^-3 \tilde{U})
= r^{-3} \Delta \tilde{U} + 2 \nabla r^{-3} cdot \nabla \tilde{U} + \Delta (r^{-3}) \tilde{U}
= r^{-3} \left[ \Delta \tilde{U} - \frac{6}{r} \tilde{U}_r \right]
= r^{-3} \left[ \tilde{U}_{rr} - \frac{2}{r} \tilde{U}_r \right]
$$

Then

$$
\tilde{U}_{tt} = \tilde{U}_{rr} - \frac{2}{r} \tilde{U}_r\\
\tilde{\tilde{U}} = \frac{1}{r} \rightarrow \tilde{U}_r = r \tilde{\tilde{U}}\\
\tilde{U} = \int r \tilde{\tilde{U}}\\
\int^{r} r \tilde{\tilde{U}} = (r \tilde{\tilde{U}})_r  - 2 \tilde{\tilde{U}}\\
r \tilde{\tilde{U}}_{tt} = (r \tilde{\tilde{U}})_{rr} - 2 \tilde{\tilde{U}}_r = r \tilde{\tilde{U}}_{rr} + 2 \tilde{\tilde{U}}_r - 2 \tilde{\tilde{U}}_r\\
\tilde{\tilde{U}}_{tt} = \tilde{\tilde{U}}_{rr}
$$

$$
U_{tt} = U_{rr} + \frac{4}{r} U_r\\
U = r^{-3} \tilde{U} \rightarrow \tilde{U} = r^3 U\\
\tilde{\tilde{U}} = \frac{1}{r} \tilde{U}_r = \frac{1}{r} (r^3 U)_r\\
$$

Then, since $n = 2 k + 1$ for some $k$, we have
$$
\left( \frac{1}{r} \frac{\partial}{\partial r} \right)^{k - 1} [r^{n - 2} U]
$$
solves the 1D wave equation, which we can use to solve the 5D wave equation

### Even dimensions? Use Dirichlet's Principle

Fix $s > 0$, then solve

$$
u_{tt} = \Delta u, x \in R^n, t > s\\
u(x, s; s) = 0, u_t(x, s; s) = f(x, s)
$$

Call this solution $U(x, t, s)$ with $t > s$

## Theorem
$u(x, t) = \int_0^t U(x, t; s) ds$
This solves the wave equation
$$
\begin{cases}
u_{tt} = \Delta u + f(x, t)\\
u(x, 0) = 0, u_t(x, 0) = 0
\end{cases}
$$

### Proof

$$
u(x, 0) = 0 = \int_0^0 U_t(x, t; s) ds\\
u_{tt}(x, t) = U_t(x, t; t) + \int_0^t U_{tt}
= f(x, t) + \int_0^t \Delta_x U\\
= f(x, t) + \Delta_x \int_0^t U\\
= f(x, t) + \Delta_x u
$$
