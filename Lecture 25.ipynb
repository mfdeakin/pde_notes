# Part 3: Weak solutions
$$
\begin{cases}
-\partial_i (a^{i, j} \partial_j u) + b_j \partial_j u + c u = f & \in \Omega\\
u = 0 & \in \partial \Omega
\end{cases}
$$

Aim: Find a weak solution
$$
a^{i, j} \zeta_i \zeta_j \geq \theta |\zeta|^2, a^{i, j} = a^{j, i} \in L^\infty, b_j, c \in L^\infty\\
B[u, v] = \int_\Omega [\a^{i, j} \partial_i u \partial_j v + b_j \partial_j u v + c u v] \dx
$$
$B$ is a bilinear form
## Theorem 1 (Lax-Milgram)
$B: H \times H \rightarrow R$, bilinear satisfies
1) Boundedness: $|B[u, v]| \leq C ||u|| ||v||
2) Coerciveness: B[u, v] \geq C_1 ||u||^2

Then $\forall f \in H^*$, $\exists$ a unique $u$ s.t. $B[u, v] = <f, v>, \forall v \in H$.

### Proof
1) Fix $u \in H$.
$B[u, \cdot]: H \rightarrow R$
Then by Riesz Represntation Theorem, we have a unique $A(u) \in H$ s.t. $B[u, v] = <A(u), v>_H$ $\forall v \in H$
$A$ satisfies
1) $A: H \rightarrow H$

2) $A$ is linear

3) $A$ is bounded

Then $|B[u, v]| = |<A(u), v>| \leq C ||u|| ||v||$, and $||A(u)|| = \sup \frac{<A(u), v>}{||v||} \leq C ||u||$

4) $A$ is one to one
If $A(u) = 0$, $B[u, v] = 0$ $\forall v \in H$, and $C_1 ||u||^2 \leq B[u, u] = 0 \rightarrow u = 0$
Then $\text{range}(A) \subset H$ is closed

5) $A$ is onto
$H = R(A) \cplus (R(A))^\perp$
$\exists v \in (R(A))^\perp \rightarrow <A(u), v> = 0 \forall u$
$B{u, v] = <A(u), v>_H = 0 \forall u \in H, B[v, v] = <A(v), v>_H = 0 \rightarrow ||v|| = 0 \rightarrow v = 0$

Thus $A$ is a bijection from $H$ to $H$

$\forall f \in H^*$, by Riesz, we have a unique $w \in H$ s.t. $<w, v>_H = <f, v>$.
Choose $u$ s.t. $A u = w \rightarrow B[u. v] = <f, v>$

### Proof of Coerciveness (Energy Estimates)

$$
B[u, v] = \int_\Omega (a^{i, j} \partial_i u \partial_j u + b_j (\partial_j u) u + c u^2) dx\\
\int_\Omega a^{i, j} \partial_i u \partial_j u \geq \theta ||D u||_{L^2}^2\\
$$
$$
\left| \int_\Omega b_j (\partial_j u) u dx \right| \leq ||b_j||_{L^\infty} ||D u||_{L^2} ||u||_{L^2}||\\
$$
$$
\left| \int_\Omega c u^2 dx \right| \leq ||c||_{L^\infty} ||u||_{L^2}^2
$$
Thus
$$
B[u, u] \geq (\theta - \epsilon ||b||_{L^\infty}) ||D u||_{L^2}^2 - \left( \frac{||b||_{L^\infty}}{4 \epsilon} + ||c||_{L^\infty} \right) ||u||_{L^2}^2
$$

Choose $\epsilon$ s.t. $||b||_{L^\infty} \epsilon = \frac{\theta}{2}$
$$
B[u, u] \geq \frac{\theta}{2} ||D u||_{L^2}^2 - \gamma ||u||_{L^2}^2\\
B[u, u] \geq c_1 ||u||_H^2 - \gamma ||u||_{L^2}^2
$$
completing our proof

Consider $B_\mu[u, v] = \int_\Omega (a^{i, j} \partial_i u \partial_j u + b_j (\partial_j u) u + c u^2) dx + \mu u v$:
Then for $\mu \geq \gamma$, this new Bilinear form satisfies boundedness and coerciveness by the Lax-Milgram theorem

## Theorem 1 (Existence Theorem 1)
$$
\forall f \in H^{-1}(\Omega), \mu \geq \gamma
$$
We have a unique weak solution to
$$
\begin{cases}
-\partial_i (a^{i, j} \partial_j u) + b_j \partial_j u + c u + \mu u = f & \in \Omega\\
u = 0 & \in \partial \Omega
\end{cases}\\
||u||_H \leq C ||f||_{H^{-1}}
$$

## Corollary
$f \in L^2(\Omega) \rightarrow \forall \mu \geq \gamma$, there exists a unique weak solution $u$ s.t.
$$
\begin{cases}
-\partial_i (a^{i, j} \partial_j u) + b_j \partial_j u + c u + \mu u = f & \in \Omega\\
u = 0 & \in \partial \Omega
\end{cases}\\
||u||_{H_0^1(\Omega)} \leq C ||f||_{L^2(\Omega)}\\
C = C(a^{i,j}, b_j, c, \mu, n, \Omega)
$$
Formally, $(L + \mu)[u] = f, u = (L + \mu)^{-1}[f]$

How does this translate into solving our original problem?
$L[u] = f$?

We have the new solvable problem $L[u] + \mu u = f$ for $f \in L^2$ we have a unique $u \in H_0^1(\Omega)$ s.t. $L[u] + \mu u = f$ holds, with $||u||_{H_0^1} \leq C ||f||_{L^2}$ and $f \rightarrow K[f] = u$

We write the original problem as
$$
L[u] + \mu u = \mu u + f \leftarrow \rightarrow u = K[\mu u + f] \leftarrow \rightarrow (I - \mu K)[u] = K[f]
$$
Call $g = K[f], T = -\mu K, \leftarrow \rightarrow (I + T)(u) = g$.

$T$ is called a compact map from $L^2$ to $L^2$

Recall the following functional analysis result:
There are many ways to define compactness, we just define it as sequential compactness
Let $H$ be a Hilbert space. We have a map $T: H \rightarrow H$, which is called compact if any bounded sequence $\{u_m\}$ has at least one convergent subsequence of $\{T(u_m)\}$

## Fredholm Alternatives
Suppose $T$ is compact. Then either $\forall g \in H$, there is a unique $u$ s.t. $(I + T)(u) = g$ and $g \rightarrow u$ is bounded, or $\exists u_0 \in H$, $u_0 \neq 0$, s.t. $(I + T)(u_0) = 0$, which is called the kernel of the operator.

In the second case, $dim(I + T^*) < \infty$ and $\forall g \perp \{u_0 | (I + T^*)(u_0) = 0 \}$, there is a solution to $(I + T)(u) = g$
