# Moser's Lemma
$$
\begin{cases}
-\partial_i (a^{i, j} \partial_j u) + c(x) u = f & x \in \Omega\\
u = 0 & x \in \partial \Omega
\end{cases}
$$

## Weak solution
$$
\int a^{i, j} \partial_i u \partial_j v + c(x) u v dx = \int f v\\
\forall v \in H_0^1(\Omega)
$$

## Weak sub-solution
$$
\int a^{i, j} \partial_i u \partial_j v + c(x) u v \leq \int f v\\
\forall v \in H_0^1(\Omega)
$$
Weak super-solution if $\geq$ rather than $\leq$
Weak solution $\leftarrow \rightarrow$ weak sub-solution, weak super-solution

## Theorem 1
1) $u \in H^1(\Omega)$ is a weak sub-solution

2) $c(x), f(x) \in L_{loc}^q(\Omega)$, $q > \frac{n}{2}$

Then $\sup_{B_{\frac{1}{2}}} u \leq c ( ||u^+||_{L^2(B_1)} + ||f||_{L^q(B_1)} )$
As a consequence, $u \in L_{loc}^{\infty}(\Omega)$

One application
$$
\begin{cases}
-\Delta u = \lambda |u|^{p - 1} u & x \in \Omega\\
u = 0 & x \in \partial \Omega
\end{cases}
$$

$u$ is a weak solution $\leftarrow \rightarrow \int_\Omega \nabla u \cdot \nabla v = \lambda \int_\Omega |u|^{p - 1} u v$ $\forall v \in H_0^1(\Omega)$

### Claim
If $p < \frac{n + 2}{n - 2}$ for $n \geq 3$ or $p < \infty$ if $n = 2$, then $u \in C^{2, \alpha}$

#### Proof
$$
c(x) = \lambda |u|^{p - 1}\\
u \in H_0^1(\Omega) \rightarrow u \in L^{\frac{2 n}{n - 2}}(\Omega)\\
\int_\Omega (|u|^{p - 1})^q = \int_{\Omega} |u|^{(p - 1) q} < \infty \if (p - 1) q = \frac{2 n}{n - 2}\\
q = \frac{2 n}{n - 2} \frac{1}{p - 1} > \frac{2 n}{n - 2} \frac{1}{\frac{n + 2}{n - 2} - 1} = \frac{n}{2}\\
c(x) \in L^{q}(\Omega), q > \frac{n}{2}
$$
By Moser's Lemma, $u \in L_{loc}^{\infty}(\Omega)$
By Sobolev embedding
$$
\lambda |u|^{p - 1} u \in L_{loc}^{\infty}(\Omega) \rightarrow \lambda |u|^{p - 1} u \in L_{loc}^q(\Omega), q > 1
$$
Then by $W^{2, q}$ embedding theory, $u \in W_{loc}^{2, q}(\Omega)$, $\forall q > n \rightarrow u \in C_{loc}^{1, \alpha} \rightarrow |u|^{p - 1} u \in C^{1, \alpha}$

Then Moser's Lemma acts a s a bridge
