# Max/Min Principle
If
$$
\begin{cases}
a^{i, j} u_{x_i, x_j} + b_{j} u + c(x) u \geq 0 & \in \Omega\\
c(x) \leq 0 & \in \Omega
\end{cases}
$$
Then $u$ can't attain it's non-negative max unless $u = c$.

We are able to replace $c(x) \leq 0$ by either $M = 0$ or $\exists$ a super solution

## Applications of the Maximum principle
1) A-priori estimates

2) Gradient Estimates (Bernstein Estimates)

3) Harnack Inequality

$$
\begin{cases}
a^{i, j} u_{x_i, x_j} + b_{j} u_{x_j} + c(x) u = f & \in \Omega\\
u = g & \in \partial \Omega
\end{cases}
$$

### Theorem 1
Suppose $c(x) \leq 0$ in $\Omega$. Then
$$
\max_{\bar{\Omega}} |u| \leq C (\max_{\bar{\Omega}} |f| + \max_{\partial \Omega} |g|)\\
C = C(\Omega, (a^{i, j}), b_j, c(x))
$$
So you also get uniqueness

#### Proof
Idea: Try to find a smooth positive solution $w$ s.t. $L w \leq -F$ in $\Omega$ and $w \geq G$ on $\partial \Omega$, where $F = \max_{\bar{\Omega}} |f|$, $G = \max_{\partial \Omega} |g|$.
$$
\begin{cases}
L (u - w) = f - L w \geq f + \max |f| \geq 0 \rightarrow \max_{\bar{\Omega}} (u - w) = \max_{\partial \Omega} (u - w)\\
u - w = g - w \leq g - \max_{\partial \Omega} |g| \leq 0 \rightarrow u \leq w
\end{cases}
$$
Also, $-u \leq w$, so $|u| \leq w$

Consider $\Omega \subset \{ 0 < x_1 < d \}$. Then the maximum principle applies due to the narrowness condition.
Then try
$$
w(x) = G + C (e^{\alpha d} - e^{\alpha x_1})\\
L w = c(x) G + (-C \alpha^2 a^{1, 1} - C b^{1} \alpha) e^{\alpha x_1} + c(x) C (e^{\alpha d} - e^{\alpha x_1})
$$
The first term and the last term are bounded above by zero (ie, they're non-positive). Then
$$
L w \leq -C (\alpha^2 a^{1, 1} + b^{1} \alpha) e^{\alpha x_1}
$$
Just choose $\alpha$ large enough s.t. $\alpha^2 a^{1, 1} + b^{1} \alpha \geq 1$, then take $C = F$.
We have
$$
L w \leq -C (\alpha^2 a^{1, 1} + b^{1} \alpha) e^{\alpha x_1} \leq -F
$$
Thus, it's non-positive as well
$$
w = G + F(e^{\alpha d} - e^{\alpha x_1})\\
|u| \leq w
$$

### Theorem 2
Suppose $c(x) \leq 0$ in $\Omega$, $\alpha(x) \geq \alpha_0 > 0$.
Then we again have
$$
\max_{\bar{\Omega}} |u| \leq C (\max_{\bar{\Omega}} |f| + \max_{\partial \Omega} |g|)\\
C = C(\text{diam}(\Omega), \theta, b_j, \alpha_0)
$$

#### Proof
$$
w = \frac{1}{\alpha_0} F + \frac{1}{\alpha_0} G\\
u(x) = w(x) v(x)
L (u - w) = f - c(x) w
$$


Idea $c(x) + \frac{L w}{w} < 0$, $\frac{1}{w} \frac{\partial w}{\partial \nu} + \alpha(x) \geq 0$
Claim:
$$
\tilde{v} \leq \frac{\max \tilde{f}}{\min |\tilde{c}|} + \frac{\max \tilde{g}}{\min |\tilde{\alpha}|}\\
\max_{\bar{\Omega}} \tilde{v} = \tilde{v}(x_0)
$$
If $x_0 \in \Omega$, $a^{i, j} v_{i, j} \leq 0$, $b_j v_j = 0$, $\tilde{c}(x) v \geq \tilde{f}$, then
$$
\rightarrow v \leq \frac{\tilde{f}}{-\tilde{c}}\\
x_0 \in \partial \Omega, \text{HBL} \rightarrow \frac{\partial v}{\partial nu} \leq 0 \rightarrow \tilde{a} v \leq \tilde{g}
$$

## Gradient Estimates (Bernstein Estimates)

$$
L u = a^{i, j} u_{x_i, x_j} + b^{j} u_{x_j} + c(x) u = f \hspace{5mm} \in \Omega
$$

### Theorem 3
$\max_{\bar{\Omega}} |\nabla u| \leq C (\max_{\partial \Omega} |\nabla u| + \max_{\bar{\Omega}} |u| + \max_{\bar{\Omega}} |f|)$

#### Proof
$$
L (|\nabla u|^2)\\
|\nabla u|^2 = u_k u_k\\
(|\nabla u|^2)_{x_i} = (u_{x_k} u_{x_k})_{x_i} = 2 u_{x_k} u_{x_k, x_i}\\
(|\nabla u|^2)_{x_i, x_j} = (2 u_{x_k} u_{x_k, x_i})_{x_j} = 2 (u_{x_k, x_j} u_{x_k, x_i} + u_{x_k} u_{x_k, x_i, x_j})\\
a^{i, j}(|\nabla u|^2)_{i, j} = 2 a^{i, j} u_{x_k, x_j} u_{x_k, x_i} + 2 a^{i, j} u_{x_k} u_{x_k, x_i, x_j})\\
(a^{i, j} u_{x_i, x_j} + b^{j} u_{x_j} + c u)_{x_k} = f_{x_k}\\
a^{i, j} u_{x_i, x_j, x_k} + a^{i, j}_{x_k} u_{x_i, x_j} + b^{j}_{x_k} u_{x_j} + b^{j} u_{x_j, x_k} + c_{x_k} u + c u_{x_k} = f_{x_k}\\
  \geq \theta |u_{x_k, x_i}|^2 - 2 u_{x_k} (a^{i, j}_{x_k} u_{x_i, x_j} + b^{j}_{x_k} u_{x_j} + b^{j} u_{x_j, x_k} + c_{x_k} u + c u_{x_k} + f_{x_k}) \geq \theta |u_{x_i, x_j}|^2
$$
Incomplete...

This is a global estimate, next we see a local estimate which is much more useful
