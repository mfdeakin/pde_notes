# Part 3 Weak Solutions
$$
\begin{cases}
-\partial_i(a^{i, j} \partial_j u) + b_j \partial_j u + c(x) u = f & \in \Omega\\
u = 0 & \in \partial \Omega
\end{cases}\\
f \in L^2(\Omega)\\
L[u] = -\partial_i(a^{i, j} \partial_j u + b_j \partial_j u + c(x) u
$$

## Theorem 1 (Existence 1)
$\exists \gamma > 0$ s.t. $\forall \mu \geq \gamma$
$$
\begin{cases}
L[u] + \mu u = f & \in \Omega\\
u = 0 & \in \partial \Omega
\end{cases}
$$
has a unique weak solution $||u||_{H_0^1(\Omega)} \leq C ||f||_{L^2(\Omega)}$

Fix $\mu = \gamma$
Then
$f \in L^2 \rightarrow \exists$ a unique $u = (L + \gamma)^{-1}[f] = K_{\gamma}[f]$

$||K_\gamma[f]||_{H_0^1} \leq c ||f||_{L^2}$

We write
$$
L[u] = L[u] + \gamma  u - \gamma u = f\\
L[u] = f \leftarrow \rightarrow u = K_\gamma[\gamma u + f]\\
\leftarrow \rightarrow (I - T)[u] = g\\
T = \gamma K_{\gamma}, g = K_{\gamma}[f]\\
T: L^2 \rightarrow H_0^1 \subset L^2
$$
Instead consider $T: L^2 \rightarrow L^2$.

## Lemma 1
$$
T: L^2 \rightarrow L^2
$$
is a compact operator.

### Proof
Compact operators means that a bounded sequence should have a convergent subsequence.

Consider $||u_m||_{L^2} \leq 1$ Then $||T[u_m]||_{H_0^1} \leq C ||u_m||_{L^2} \leq C$ by Theorem 1.

Since $H_0^1 \subset \subset L^2$ (compact embedding), $T[u_m]$ must have a convergent subsequence, proving compactness.

We have the identity map minus compact map $(I - T)[u] = g$
##### Then by a theorem from functional analysis:
Let $T: H \rightarrow H$ be a compact operator. Then we can drop out the Hilbert operator, and the following holds: Either $\forall g \in H$, $\exists$ a unique $u \in H$ s.t. either $u - T[u] = g$ or $u_0 \neq 0, u_0 - T[u_0] = 0$.
##### This is the Fredholm Alternative
One of the basic tools we use to solve PDEs.

In case 1, $g \rightarrow u = (I - T)^{-1}[g]$ is bounded.

In case 2, $dim\{u_0 | u_0 - T[u_0] = 0\} < \infty$, and $u - T[u] = g$ is solvable iff $(g, v) = 0$ for any $v \in \{ u_0 | u_0 - T^*[u_0] = 0 \}$, where $(T u, v) = (u, T^* v)$, $T^*$ is the conjugate operator or dual operator.

$H = \{ u_0 | u_0 - T[u_0] = 0 \} + \left( \{ u_0 | u_0 - T u_0 = 0 \} \right)^{\perp}$ (which you are in if the inverse is not zero)
and you can also decompose it to
$H = \{ u_0 | u_0 - T^*[u_0] = 0 \} + \left( \{ u_0 | u_0 - T^*[u_0] \} \right)^{\perp}$

Conjugate operator
$T[f] = \gamma K_\gamma[f]$
$L = -\partial_i (a^{i, j} \partial_j u) + b_j \partial_j u + c(x) u$

If I look at the $L^2$ norm, integrating by parts gives $(L u, v) = (u, L^* v)$.
$$
\int b_j (\partial_j u) v = -\int u \partial_j (b_j v) = -\int u (\partial_j v) b_j - \int u (\partial_j b_j) v\\
L^* v = -\partial_i (a^{i, j} \partial_j v) - b_j \partial_j v + (c(x) - \partial_j b_j) v
$$
Applying the Fredholm Alternative to
$u - \gamma K_\gamma[u] = g$ implies either $\forall g \in L^2$, $\exists$ a unique $u \in L^2$ s.t.
$u - \gamma K_\gamma[u] = g$ or $\exists u_0 \in L^2$ s.t. $u_0 - \gamma K_\gamma[u_0] = 0$

Our situation is $u - \gamma K_\gamma[u] = K_\gamma[f]$
In case 1
$g = K_\gamma[f] \rightarrow \exists$ a unique $u \in L^2$ s.t. $u - \gamma K_\gamma[u] = K_\gamma[f]$
I have a problem, $g$ is only in $L^2$, not in $H^1_0$.
But since $K_\gamma[f], K_\gamma[u] \in H_0^1$, we must have $u \in H_0^1$, meaning that it is a weak solution,
with $L[u] = f$

In the second case, I found
$\exists u_0 \in L^2$ s.t. $u_0 - \gamma K_\gamma[u_0] = 0$
Then $u_0 - \gamma K_\gamma[u_0] \in H_0^1$, so $u_0$ is a weak solution

We are using the regularity theorems

## Theorem 2 (Existence Theorem 2)
Either $\forall f \in L^2$, there is a unique weak solution to $L[u] = f$ in $\Omega$ with $u = 0$ in $\partial \Omega$.
We also have $||u||_{H_0^1} \leq c ||f||_{L^2}$.
Or $\exists u_0 \neq 0$, $u_0$ is a weak solution to the homogeneous problem
$$
\begin{cases}
L[u_0] = 0 & \in \Omega\\
u_0 = 0 & \in \partial \Omega
\end{cases}
$$

In this case, we can still solve the first problem iff $\int_\Omega f v = 0 \forall v$ which satisfy
$$
%\begin{cases}
%L u = f & \in \Omega\\
%u = 0 & \in \partial \Omega
%\end{cases}
\begin{cases}
L^* v = 0 & \in \Omega\\
v = 0 & \in \partial \Omega
\end{cases}
$$

### Remark
In case two, you can find a unique solution $u$ with $u \perp$ any $u_0$ with
$$
\begin{cases}
L u_0 = 0 & \in \Omega\\
u_0 = 0 & \in \partial \Omega
\end{cases}
$$

In case 2, $dim(\{L u_0 = 0\}) = dim(\{L^* u_0 = 0\}) < \infty$
$$
L^2 = \{ L u_0 = 0 \} + \{ L u_0 = 0 \}^\perp\\
L^2 = \{ L^* u_0 = 0 \} + \{ L^* u_0 = 0 \}^\perp
$$

This is a complete solution to the original problem. We don't need to assume that the operator is self adjoint or anything

$u_0 \ in L^2$, $u_0 - \gamma K_\gamma[u_0] = 0 \rightarrow u_0 \in H_0^1 \rightarrow u_0$ is a weak solution to
$$
\begin{cases}
L u_0 = 0 & \in \Omega\\
u_0 = 0 & \in \partial \Omega
\end{cases}
$$
Case 2 has a solution iff $\int_\Omega K_\gamma[f] v dx = 0$ whenever $v - (\gamma K_\gamma)^*[v] = 0 \leftarrow \rightarrow \int_\Omega f v = 0$
$$
K_\gamma[f] = u \leftarrow \rightarrow L u + \gamma u = f\\
\int(L u, v) + \gamma \int u v = \int f v = \int(u, L^* v) + \gamma \int u v = \int f v \rightarrow \int_\Omega f v = 0\\
\int K_\gamma[f] v = 0 \leftarrow \rightarrow (K_\gamma[f], v) = (f, K_\gamma^*[v]) = (f, \frac{1}{\gamma} v) = \frac{1}{\gamma}(f, v)
$$
