## Theorem 2
$u \in C_0^\infty(R^n)$

$||u||_{L^{\frac{n p}{n - p}}(R^n)} \leq C_{n, p} || D u||_{L^p(R^n)}$

We know (but do not prove in this class) that
$C_{n, p}$ is attained iff $u = \left(1 + \frac{|x|^2}{n (n - 2)} \right)^{(p - n) / 2}$

### Proof
For $p = 1$

$$
u(x_1, x_2, \ldots, x_n) = \int_{-\infty}^{x_1} D_1 u(t, x_2, \ldots, x_n dt\\
|u(x_1, x_2, \ldots, x_n)|^{\frac{n}{n - 1}} \leq
  \left( \int_{-\infty}^{\infty} |D_1 u| dx_1 \int_{-\infty}^{\infty} |D_2 u| dx_2 \ldots \int_{-\infty}^{\infty}
|D_n u| dx_n \right)^{\frac{1}{n - 1}}\\
|u(x_1, x_2, \ldots, x_n)|^{\frac{n}{n - 1}} \leq
  \left( \int_{-\infty}^{\infty} |D_1 u| dx_1 \right)^{\frac{1}{n - 1}} \left( \int_{-\infty}^{\infty} |D_2 u| dx_2 \ldots \int_{-\infty}^{\infty} |D_n u| dx_n \right)^{\frac{1}{n - 1}}\\
$$

Can use the Cauchy inequality to get
$$
\leq
  \left( \int_{-\infty}^{\infty} |D_1 u| dx_1 \right)^{\frac{1}{n - 1}}\\
